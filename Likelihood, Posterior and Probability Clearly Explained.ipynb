While the probability of drawing _one_ data point from a model with parameters $\theta$ **equals** the _likelihood_ of the parameters of a model given a data point:

$$
\boxed{
P(x_i|\theta) = \mathcal {L}(\theta | x_i)
}
$$

the integral of a PDF (PMF in the random variable is discrete) and the integral of a likelihood function (which is always continuous) are **not the same**:

$$
\boxed{
\sum_i P(x_i|\theta) \neq \int \mathcal {L}(\theta | x_i)  d\theta
}
$$

*Note*: we sum over the data points (while keeping the parameters constant, but integrate over the parameters, keeping the data constant.

This is because while the outcomes (the data, $x_i$) are _discrete_ and _mutually exclusive_ things and there is a limited space for them, we can integrate them to 1. But the hypotheses (or parameters) $\theta$ _can_ be mutually inexclusive (nested) and there's a potentially unlimited space of hypotheses (parameters) you can make about the data(set) ($\mathcal L(\theta|X)$, the likelihood doesn't have to (and almost never) integrate to 1.

In plain Engish: the probability of a data point given the parameters of model (distribution) $\theta$ equals the likelihood of the parameter given that datapoint.

**IMPORTANT**

with the _probability_ you _hold the parameters constant_ and just sample from the model (in other words, you take a model parametrized by $\theta$ and use it to generate data. This is a generative model.

with the _likelihood_ you _hold the data constant_ and _only vary the parameter_ $\theta$. The crucial difference between probability and likelihood is that unlike probability, likelihood must not integrate to 1.

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.random.randn(20)
plt.scatter(X, np.ones_like(X))

def pdf_norm(x, mu, sigma):
    return 1/(2*np.sqrt(np.pi))*np.exp(-0.5*((x-mu)/sigma)**2)

sigma = 1
mu = np.linspace(-3,3,100)


L = [np.log(pdf_norm(X, mu, sigma)).sum() for mu in mu]
plt.plot(mu, L)
plt.xlabel('$\mu$', fontsize=20)
plt.ylabel('$L( \mu|D)$', fontsize=20)

```

Now we have some data and we want to ask the question: given this data what would be the most likely value of parameter $\theta$ of model $m$ that this data most like came from (was generated by/ was sampled from? This is called **maximum likelihood estimate for parameter $\theta$**. For simplicity we will only try do MLE for the mean of the normal distribution (model m). So we will take a vector of values of $\theta$ and record the values of the likelihood. 

The likelihood of one value of parameter $\theta$ given only one data point $x_i$, i.e. $L(\theta|x_i)$, is simply the height (value) of the model's probability density function at that data point. For example, if $x_i$ = 1, then L = 0.17109 for $\mu$ = 0 and $\sigma = 1$

$$L(\mu|x_i) = \mathcal{N}(\mu, \sigma) = {\frac {1}{\sigma {\sqrt {2\pi }}}}e^{-{\frac {1}{2}}\left({\frac {x-\mu }{\sigma }}\right)^{2}}$$



`pdf_norm(1, 0, 1) = 0.17109` (This is the PDF of our model, m, and it is defined by the Gaussian function)

But often we have not one data point, but a big dataset of points $D = \{x_1,...x_n\}$. In that case, we need to estimate $L(\theta|\{x_1,...x_n\})$, which (by the product rule of probabilities, is $\prod_i L(\theta|x_i)$. However, there's an issue with this formula: in practice multiplication requres all datapoints at once, and is not practical for large datasets and stochastic optimization. So instead of likelihood, **log-likelihood** is used (becase the log of a product is the sum of logs).

$$
\log L = \sum_i \log L(\theta|x_i)
$$

In the graph above we have a bunch of data points drawn from a distribution (generated by a generative model) with parameter $\mu = 0$, and we plot the log-likelihood computed for different values of parameter $\mu$. The value $\mu$ where the log-likelihood reaches a maximum is the parameter of the distribution (model) where the data MOST LIKELY have been drawn from.

**Likelihood**:
$$
\mathcal L(\theta|x)
$$
Likelihood of what? Likelihood of the model, given the data.


Model is an equation that best describes the observed data. It can be defined and as Gaussian function, a sinusoid or any other function.
$$
\theta = \{\mu, \sigma\}
$$


**Probability**:
$$
P(x|\theta)
$$
Probability of what? Probability of data given the model.


**With probability:**

We are given a model and its parameters (and they are fixed) and we draw data/samples from it. A certain distribution of these results/outcomes is obtained. The integral of probabilities is 1.

**With likelihood:**

We are given data (and we don’t draw them, we already have them). Now we try different models and parameters (i.e. different (m, 𝜃)) and see how likely they are given the data.

Probabilities attach to results; likelihoods attach to hypotheses.


# terminology
## Likelihood
Not to confuse likelihood with posterior, look out for data on the first place:
$$P(x|\theta)$$
Note that likelihood can also be written as 
$$\mathcal L(\theta | x)$$
which is pretty confusing, because the order of $x$ and $\theta$ has changed! 

## Posterior
Posterior (posterior estimate of an unobserved quantity / parameter _conditional_ on the data):
$$P(\theta | x)$$

## Comlete data likelihood
If you have more than one predictor in the data, your likelihood can look like this:

$$P(y, w|s, b, \lambda) = P(y|s, w, b, \lambda)p(w)$$

## Marginal likelihood
If you have more than one predictor in the data, BUT you marginalize out one of them, your likelihood can look like this:

$$P(y|s, b, \lambda)$$

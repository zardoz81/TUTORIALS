```python
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy.stats as stats

mu = 0
variance = 1

sigma = math.sqrt(variance)
x = np.linspace(-5, 5, 100)
pz = stats.norm.pdf(x, mu, sigma)
plt.plot(x, pz, label='prior')
plt.axvline(0, linewidth=0.5, linestyle = '-.')


mu = 1
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(-5, 5, 100)
px = stats.norm.pdf(x, mu, sigma)
plt.plot(x, px, label='evidence')


mu = 0.5
variance = 1
pred = stats.norm.pdf(x, mu, sigma)
plt.plot(x, pred, label='likelihood')

pzx = pred*pz/px
pzx /= np.trapz(y=pzx, dx=np.diff(x).min(), axis=-1) # normalized to get a distribution

plt.plot(x, pzx, label='posterior') 

plt.legend()
```

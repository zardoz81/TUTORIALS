```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import torch
from torch.autograd import Variable
import torch.optim as optim


# make a few points and plot them:
a = np.array([1, 3, 6, 3, 9, 5, 7, 3, 7, 2, 5, 3, 4, 5]).flatten()
b = np.array([3, 6, 1, 8, 3, 7, 3, 7, 0, 9, 6, 8, 1, 4]).flatten()

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.scatter(a,b)
plt.title('Original points')
for i in range(14):
    plt.annotate(i, (a[i], b[i]))


def get_d(x, i, j):
    # compute distances between ith and jth point:
    d = (x[0,i] - x[0,j])**2 + (x[1,i] - x[1,j])**2
    return d if d==0 else d.pow(0.5)


# put the real points into a torch tensor:
x = torch.stack((torch.tensor(a, dtype=torch.float), torch.tensor(b, dtype=torch.float)))

# make some random points that we'll be moving through SGD:
x_ = Variable(torch.rand((2,14)),  requires_grad=True)


LOSS = []
for epoch in range(100):
    epoch_loss = torch.tensor([0], dtype=torch.float)
    for i in range(14):
        for j in range(14):
            # compute the squared error between the distance beteen real point and between generated points: 
            epoch_loss += (get_d(x,i,j)-get_d(x_,i,j)).pow(2)
    # compute the gradient:
    epoch_loss.backward(retain_graph=True)
    # move the points aganst the gradient:
    x_.data -= 0.01 * x_.grad.data
    # zero the gradient (because it accumulates):
    x_.grad.data.zero_()
    LOSS.append(epoch_loss.item())
    
plt.subplot(1,2,2)
plt.title('Loss')
plt.plot(LOSS)

plt.figure()
xx = x_.detach().numpy()
plt.scatter(xx[0,:], xx[1,:])
plt.title('Random points optimized thru SGD')
for i in range(14):
    plt.annotate(i, (xx[0,i], xx[1,i]))
    
```

## Now we'll do it through PCA

https://math.stackexchange.com/questions/156161/finding-the-coordinates-of-points-from-distance-matrix

$$
M_{ij} = \frac{D^2_{1j} + D^2_{i1} - D^2_{ij}} {2}
$$

## For derivation see Young and Householder 1938 (Pdf in this folder)

![and my workthrough](./img/RSA.jpg)

```python
D = np.zeros((14,14))
for i in range(14):
    for j in range(14):
        D[i,j] = get_d(x,i,j).detach().numpy()
```

```python
u,s,v = np.linalg.svd(D)
```

```python
# the rank of the matix
D.ndim
```

```python
M = np.zeros_like(D)
for i in range(14):
    for j in range(14):
        M[i,j] = 0.5 * (D[0, j]**2 + D[i,0]**2 - D[i,j]**2);
```

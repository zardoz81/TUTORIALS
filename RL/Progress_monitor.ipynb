```python
import matplotlib.pyplot as plt
import time
import numpy as np
%matplotlib inline



file = "log_PG.txt"
f=open(file,"r")
lines=f.readlines()
loss, epsilon, reward =[],[],[]
for x in lines:
    loss.append(np.array(x.split(' ')[2], dtype='float32'))
    epsilon.append(np.array(x.split(' ')[4], dtype='float32'))
    reward.append(np.array(x.split(' ')[6], dtype='float32'))
f.close()

plt.subplot(3,1,1)
plt.plot(loss)
plt.title('loss')
plt.subplot(3,1,2)
plt.plot(epsilon)
plt.title('sum of raw discounted rewards')
plt.subplot(3,1,3)
plt.plot(reward)
plt.title('sum of raw UNdiscounted rewards')
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
import time
import numpy as np
%matplotlib inline


l = []
file = "log_PG_1.txt"
file = "2_PG_4frames.txt"
f=open(file,"r")
lines=f.readlines()
loss, epsilon, reward =[],[],[]
for x in lines:
    l = ''.join(e for e in x.split(' ')[2] if e.isdigit())
    loss.append(np.array(l, dtype='float32'))
    e = ''.join(e for e in x.split(' ')[4] if e.isdigit())
    epsilon.append(np.array(e, dtype='float32'))
    r = ''.join(e for e in x.split(' ')[6] if e.isdigit())
    reward.append(np.array(r, dtype='float32'))
f.close()

plt.subplot(3,1,1)
plt.plot(loss)
plt.title('loss')
plt.subplot(3,1,2)
plt.plot(epsilon)
plt.title('EPISODE REWARD')
plt.subplot(3,1,3)
plt.plot(reward)
plt.title('NUMBER OF DOWN ACTS')
plt.tight_layout()
```

## Derivation of Eq. 8.19a

$$
w(t) = \sum_{\mu=1}^{N_u} c_{\mu}(t)\mathbf{e}_{\mu}
$$

and if we take the dot product of both sides with $\mathbf{e}_{\mu}$, because the the product of any two different eigenvectors is zero (they are orthogonal) and one if they are the same, we can also express 
$$ c_{\mu}(t) = w(t)\mathbf{e}_{\mu} $$

$$
\tau_w \frac{dw}{dt} = \mathbf{Q} \mathbf{w}
$$

if we plug the first equation instead of $\mathbf{w}$ in the third, we get:

$$
\tau_w \frac{dw}{dt} = \mathbf{Q} \sum_{\mu=1}^{N_u} c_{\mu}(t)\mathbf{e}_{\mu}
$$

We can put $\mathbf{Q}$ under the sum sign because it doesn't depend on the index $\mu$:
$$
\tau_w \frac{dw}{dt} = \sum_{\mu=1}^{N_u} c_{\mu}(t) \mathbf{Q} \mathbf{e}_{\mu}
$$

Since $\mathbf{e}_{\mu}$ are (orthonormal) eigenvectors of the covariance matix $\mathbf{Q}$:
$$
\mathbf{Q} \mathbf{e}_{\mu} = \lambda_\mu \mathbf{e_\mu}
$$
we can get rid of the covariance matrix and write:
$$
\tau_w \frac{dw}{dt} = \sum_{\mu=1}^{N_u} c_{\mu}(t) \lambda_\mu \mathbf{e}_{\mu}
$$

Now we need to get rid of $\mathbf{e_\mu}$. To do so, we take the dot product of both sides of the equation with $\mathbf{e_\mu}$. Since $\mathbf{e_\mu}$ are orthogonal and orthonormal, the product of any two different eigenvectors is zero, while if they are the same, $\mathbf{e_\mu}\mathbf{e_\mu} = 1$

$$
\tau_w \frac{dw \mathbf{e_\mu}}{dt} = \sum_{\mu=1}^{N_u} c_{\mu}(t) \lambda_\mu \mathbf{e}_{\mu}\mathbf{e_\mu}
$$

$$
\tau_w \frac{d c_\mu}{dt} = c_{\mu}(t) \lambda_\mu
$$

Now we solve the ODE $\tau_w \frac{d c_\mu}{dt} = c_{\mu}(t) \lambda_\mu$

$$\tau_w \frac{d c_\mu}{dt} = c_{\mu}(t) \lambda_\mu$$

Divide both sides by $\tau_w$, $c_{\mu}(t)$ and multiply by $dt$
$$ \frac{1}{c_{\mu}(t)} d c_\mu = \frac{\lambda_\mu}{\tau_\mu} c_{\mu}(t) dt $$

Take an integral of both terms, remembering that constant factors (like $\frac{\lambda_\mu}{\tau_\mu}$) can be taken out of the integral
$$ \int \frac{1}{c_{\mu}(t)} d c_\mu = \frac{\lambda_\mu}{\tau_\mu} \int c_{\mu}(t) dt $$

$$ \ln |c_{\mu}(t)| + C_1 = \frac{\lambda_\mu}{\tau_\mu} t + C_2 $$

Take exponent of both sides:

$$ e^{\ln |c_{\mu}(t)| + C_1} = e^{\frac{\lambda_\mu}{\tau_\mu} t + C_2} $$

$$ e^{\ln |c_{\mu}(t)|}  e^{C_1} = e^{\frac{\lambda_\mu}{\tau_\mu} t} e^{C_2} $$

Divide both sides by $e^{C_1}$:

$$
c_{\mu}(t) = e^{\frac{\lambda_\mu}{\tau_\mu} t}C
$$

or

$$
\boxed{
c_{\mu}(t) = C e^{\frac{\lambda_\mu t}{\tau_\mu}}
}
$$

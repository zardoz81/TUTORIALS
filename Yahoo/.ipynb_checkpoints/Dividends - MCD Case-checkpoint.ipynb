```python
import numpy as np
import pandas as pd
pd.core.common.is_list_like = pd.api.types.is_list_like
from utils import get_dividends as get_divs
import matplotlib.pyplot as plt
import random
import seaborn as sns
import time

# To create datetime objects 
from datetime import datetime

# To grab stock data
import yfinance as fyf
from pandas_datareader import data as pdr
fyf.pdr_override() # <-- Here is the fix
```

```python
# Set label
stocks = [symbol, 'RUB=X']

# Set start and end dates
# start = datetime(2006, 1, 30)
# end   = datetime(2016, 6, 30)

df = pdr.get_data_yahoo(stocks, start=start, end=end)

dur_days = (df.index[-1] - df.index[0]).days
dur_years = dur_days/365
print('Duration: {} Days ({:.1f} Years)'.format(dur_days, dur_years))

del df['Adj Close']
del df['Volume']
del df['Open']
del df['Close']

for stock in stocks:
    df[stock] = df.loc[:, pd.IndexSlice[:, stock]].mean(axis=1)

del df['High']
del df['Low']

df.fillna(value=df.min(axis=0), inplace=True)
```

```python
df.head()
```

```python
def get_stock_price_on_date(df_stock, df_div, div_date_index, stock):
    return float(df_stock.iloc[np.nonzero(df_stock.index == df_div.index[div_date_index])[0][0]][stock])
# get_stock_price_on_date(df, div, 5, 'MCD')
```

```python
DATE, PRICE, DIV, USDRUR = [],[],[],[]
for i, j in enumerate(div.index):
    DATE.append(j.date().strftime("%b %d, %Y"))
    PRICE.append(get_stock_price_on_date(df, div, i, symbol))
    DIV.append(div[i])
    USDRUR.append(float(df.loc[div.index[i]]['RUB=X']))
#     print('{}\t\t{:.2f}\t{:.2f}'.format(j.date().strftime("%b %d, %Y"),
#                               get_stock_price_on_date(df, div, i, 'MCD'),
#                               div[i]))
DATE.reverse()
PRICE.reverse()
DIV.reverse()
USDRUR.reverse()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
plt.plot(DATE, PRICE)
a = plt.xticks(rotation=90)
plt.ylabel('Долларов за акцию')
plt.title('Стоимость акции', weight='bold')


plt.subplot(2,2,2)
plt.scatter(DATE, DIV)
plt.title('Дивиденды', weight='bold')
a = plt.xticks(rotation=90)
plt.ylabel('Долларов с акции')
plt.tight_layout()

plt.subplot(2,2,3)
plt.plot(DATE, USDRUR)
plt.title('Курс доллара', weight='bold')
plt.ylabel('Рублей за доллар')
a = plt.xticks(rotation=90)
plt.grid()
plt.tight_layout()


# CALCULATE CAPITAL GROWTH

start_cap_RUR = 300000
capUSD = start_cap_RUR/USDRUR[0]
CAPusd = [capUSD]
CAPrur = [start_cap_RUR] 
nstocks = capUSD/PRICE[0]

for i in range(1, len(DATE)):
    delta = nstocks*DIV[i]/PRICE[i]
    nstocks += delta
    CAPusd.append(nstocks*PRICE[i])
    CAPrur.append(nstocks*PRICE[i]*USDRUR[i])
nyears = (datetime.strptime(DATE[-1], "%b %d, %Y") - datetime.strptime(DATE[0], "%b %d, %Y")).days/365
interestUSDperYear = (CAPusd[-1] - CAPusd[0])/CAPusd[0]*100/nyears
interestRURperYear = (CAPrur[-1] - CAPrur[0])/CAPrur[0]*100/nyears

plt.subplot(2,2,4)
plt.plot(DATE, CAPusd)
plt.title('Капитал в долларах. Доходность: {:.2f}% годовых в долларах'.format(interestUSDperYear), weight='bold')
a = plt.xticks(rotation=90)
plt.ylabel('Если продать акции, получим ДОЛЛАРОВ')
plt.tight_layout()
plt.grid()


plt.figure(figsize=(15,5))
plt.plot(DATE, CAPrur)
plt.title('Капитал в рублях. Доходность: {:.2f}% годовых в рублях'.format(interestRURperYear), weight='bold')
a = plt.xticks(rotation=90)
plt.ylabel('Если продать акции, получим РУБЛЕЙ')
plt.tight_layout()
plt.grid()
```

```python
import matplotlib.pyplot as plt
import numpy as np
interest = 0.22 + 0.03 
cap = 20000
hist = []
time = 20*12
for i in range(time):
    if i%12 == 0:
        boost = 500
    elif i%6==0 and i%12 != 0:
        boost = 1000
    else:
        boost = 200
    cap += cap * (interest/12) + boost
    hist.append(cap)
    
    if cap > 1e6:
        cap -= 15000

int(cap)
plt.plot(np.linspace(0, int(time/12), len(hist)), [i/1000 for i in hist])
plt.grid()
plt.title('At exit: {:.0f} $'.format(cap))
plt.xlabel('лет')
plt.ylabel('капитал, тыс. $')
```

#### стартовый капитал: 20000 долл.
#### ожидаемый рост цены портфеля: 22% годовых
#### ожидаемый дивидендный доход: 3% годовых
#### выход на 1М долл.: через 14 лет
#### ежегодная докладка: 3500 долл.
#### ежемесячная пенисия в 15000 долл. через 14 лет

```python
import numpy as np
import matplotlib.pyplot as plt

n = 100
e = np.zeros((n,))
p1 = np.linspace(0.001, 0.999, n)
p2 = np.linspace(1-0.001, 1-0.999, n)
for i in range(n):
    e[i] = p1[i]*np.log2(p1[i]) + p2[i]*np.log2(p2[i])

plt.plot(p1,-e)
plt.xlabel('P(X=\'heads\')')
plt.ylabel('Entropy of X, H(X)')
```

When we flip a coin, maximum entropy (uncertainty) is achieved when the probability of tail(head) is 0.5.

# Entropy is smaller for more pointy distributions. That's because a pointy distribution better 'predicts' the outcome of a random variable, hence the entorpy is lower

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def normpdf(x, mu, sigma):
    return 1/(sigma*np.sqrt(2*np.pi)) * np.exp(-0.5*((x-mu)/sigma)**2)
    
x = np.linspace(-5, 5, 100)

p = normpdf(x, 0, 0.5)
H = np.sum(p * -np.log(p))
plt.plot(x, p, label='H = {:.3f}'.format(H))


p = normpdf(x, 0, 1)
H = np.sum(p * -np.log(p))
plt.plot(x, p, label='H = {:.3f}'.format(H))

p = normpdf(x, 0, 2)
H = np.sum(p * -np.log(p))
plt.plot(x, p, label='H = {:.3f}'.format(H))

p = normpdf(x, 0, 3)
H = np.sum(p * -np.log(p))
plt.plot(x, p, label='H = {:.3f}'.format(H))


plt.legend()

```

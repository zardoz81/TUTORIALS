```python
import sys
def printProgressBar (iteration, total, prefix = '', suffix = '', decimals = 1, length = 100, fill = '█'):
    """
    Call in a loop to create terminal progress bar
    @params:
        iteration   - Required  : current iteration (Int)
        total       - Required  : total iterations (Int)
        prefix      - Optional  : prefix string (Str)
        suffix      - Optional  : suffix string (Str)
        decimals    - Optional  : positive number of decimals in percent complete (Int)
        length      - Optional  : character length of bar (Int)
        fill        - Optional  : bar fill character (Str)
    """
    percent = ("{0:." + str(decimals) + "f}").format(100 * (iteration / float(total)))
    filledLength = int(length * iteration // total)
    bar = fill * filledLength + '-' * (length - filledLength)
    sys.stdout.write('\r%s |%s| %s%% %s' % (prefix, bar, percent, suffix))
    sys.stdout.flush()
    # Print New Line on Complete
    if iteration == total: 
        print()
```

```python
import numpy as np
import scipy
import pickle
# import hiddenlayer as hl

from keras.datasets import mnist
from keras.utils import to_categorical

from skimage import transform as im_tf

import matplotlib.pyplot as plt
%matplotlib inline

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.autograd import Variable

# make sure you don't hog all the video memory
import os

(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.
x_train = np.reshape(x_train, (len(x_train), 28, 28, 1))  # adapt this if using `channels_first` image data format
x_test = np.reshape(x_test, (len(x_test), 28, 28, 1))  # adapt this if using `channels_first` image data format
y_train_oh = to_categorical(y_train.reshape(-1), num_classes=10)
y_test_oh = to_categorical(y_test.reshape(-1), num_classes=10)


class Net(nn.Module):
    def __init__(self):
        nn.Module.__init__(self)
        self.conv1 = nn.Conv2d(1, 16, 3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(16, 16, 3, stride=1, padding=1)
        self.conv3 = nn.Conv2d(16, 32, 3, stride=1, padding=1)
        self.conv4 = nn.Conv2d(32, 32, 3, stride=1, padding=1)
        self.sc1 = nn.Conv2d(1, 16, 1, stride=1, padding=0)
        self.sc2 = nn.Conv2d(16, 32, 1, stride=1, padding=0)
        
        self.fc1 = nn.Linear(25088, 64)
        self.fc2 = nn.Linear(64, 10)
        torch.nn.init.xavier_uniform_(self.conv1.weight)
        torch.nn.init.xavier_uniform_(self.conv2.weight)
        torch.nn.init.xavier_uniform_(self.conv3.weight)
        torch.nn.init.xavier_uniform_(self.conv4.weight)
        torch.nn.init.xavier_uniform_(self.fc1.weight)
        torch.nn.init.xavier_uniform_(self.fc2.weight)
        
        self.optimizer = optim.SGD(self.parameters(), lr=0.01)
        self.criterion = nn.MSELoss()

    def forward(self, x):
#         print('\nx', x.shape)
        
        x0 = x.view(-1,1,28,28) #  это что-то вроде reshape(None,1,28,28)
#         print('x0', x0.shape) #[100, 1, 28, 28]
        
        x1 = self.conv1(x0)
#         print('x1', x1.shape) #[100, 16, 28, 28]
        x1_ = F.relu(x1)
#         print('x1_', x1_.shape) #[100, 16, 28, 28]
        
        x2 = self.conv2(x1_) #[100, 16, 28, 28]
#         print('x2', x2.shape)
        sc1 = self.sc1(x0) #[100, 16, 28, 28]
#         print('sc1', sc1.shape)
        
        x2_ = F.relu(x2 + sc1)
#         print('x2_', x2_.shape)

        x3 = self.conv3(x2_)
        x3_ = F.relu(x3)

        x4 = self.conv4(x3_)
        sc2 = self.sc2(x2_) #[100, 16, 28, 28]
        x4_ = F.relu(x4+sc2)

        x = x4_.view(-1, 25088)
        x = self.fc1(x)
        x = F.relu(x)
        
        x = self.fc2(x)
        x = F.softmax(x, dim=1)
        
        return x

    def update_parameters(self, lrate):
        count = 0
        for f in self.parameters():
            # .sub_(f.grad.data*lrate) - мы вычитаем производную (движемся против градиента, т.е. минимизируем ошибку)
            # ._sub(-f.grad.data*lrate) - мы прибавляем производную (движемся по градиенту, т.е. увеличиваем ошибку)
            f.data.sub_(f.grad.data*lrate)
                
    def fit(self, inputs, targets, lrate=0.01):
        # можно использовать готовый оптимизатор:
        # на каждом этапе мы генерируем новые predictions (т.к. они меняются, ибо меняются параметры)
#         predictions = self.forward(inputs)
        # cбрасываем градиенты (они накапливаются для каждого сэпла в бэтче)
#         self.optimizer.zero_grad()   # zero the gradient buffers
        # считаем ошибку по loss-функции (criterion). Кстати, loss-функция может быть разной.
#         self.loss = self.criterion(predictions, targets)
        # считаем градиент (частные производные loss-функции):
#         self.loss.backward(retain_graph=True)
        # чуть меняем веса против граиента:
#         self.optimizer.step()    # Does the update
        # смотрим веса, градиенты и ошибку (не всё, а то не уместится на экране):
#         count = 0
#         for f in self.parameters():
#             if count==0:
#                 print([torch.max(f.data), torch.max(f.grad.data), self.loss])
#             count =+ 1

        # либо можно сделать свой оптимизатор
        # на каждом этапе мы генерируем новые predictions (т.к. они меняются, ибо меняются параметры)
        predictions = self.forward(inputs)
        # считаем ошибку по loss-функции (criterion). Кстати, loss-функция может быть разной.
        self.loss = self.criterion(predictions, targets)
        # cбрасываем градиенты (они накапливаются для каждого сэпла в бэтче)
        self.zero_grad()
        # считаем градиент (частные производные loss-функции):
        self.loss.backward(retain_graph=True)
        # чуть меняем веса против градиента (кстати, можно и наоборот: будет gradient ascent):
        self.update_parameters(lrate)
        count = 0
        # inspect parameters:
#         for f in self.parameters():
#             if count==0:
#                 print([torch.max(f.data), torch.max(f.grad.data), self.loss])
#             count =+ 1
            
        return self.loss.detach().numpy()


b = Net()
```

```python
# Another SIMPLE TUTORIAL. Optimizing a simple neural net
# Here we adjust simply one parameter. This one is an overkill since we use a neural net model
# with only one neuron and no bias (so one parameter)

class Net(nn.Module):
    
    def __init__(self):
        nn.Module.__init__(self)
        self.fc = nn.Linear(1, 1, bias=False)
        self.criterion = nn.MSELoss()
        
    def forward(self, x):        
        x = x.view(-1,1) #  это что-то вроде reshape(None,1,28,28)
        x = self.fc(x)
        return x
    
    def update_parameters(self, lrate):
        count = 0
        for f in self.parameters():
            # .sub_(f.grad.data*lrate) - мы вычитаем производную (движемся против градиента, т.е. минимизируем ошибку)
            # ._sub(-f.grad.data*lrate) - мы прибавляем производную (движемся по градиенту, т.е. увеличиваем ошибку)
            f.data.sub_(f.grad.data*lrate)
    
    def fit(self, inputs, targets, lrate=0.01):
#        можно использовать готовый оптимизатор:
        self.optimizer = optim.SGD(self.parameters(), lr=0.001)
#        на каждом этапе мы генерируем новые predictions (т.к. они меняются, ибо меняются параметры)
        predictions = self.forward(inputs)
#        cбрасываем градиенты (они накапливаются для каждого сэпла в бэтче)
        self.optimizer.zero_grad()   # zero the gradient buffers
#        считаем ошибку по loss-функции (criterion). Кстати, loss-функция может быть разной.
        self.loss = self.criterion(predictions, targets)
#        считаем градиент (частные производные loss-функции):
        self.loss.backward(retain_graph=True)
#        чуть меняем веса против граиента:
        self.optimizer.step()    # Does the update
#        смотрим веса, градиенты и ошибку (не всё, а то не уместится на экране):
        count = 0
        for f in self.parameters():
            print(['Layer: ', count, torch.max(f.data), torch.max(f.grad.data), self.loss])
            count =+ 1

#         # либо можно сделать свой оптимизатор
#         # на каждом этапе мы генерируем новые predictions (т.к. они меняются, ибо меняются параметры)
#         predictions = self.forward(inputs)
#         # считаем ошибку по loss-функции (criterion). Кстати, loss-функция может быть разной.
#         self.loss = self.criterion(predictions, targets)
#         # cбрасываем градиенты (они накапливаются для каждого сэпла в бэтче)
#         self.zero_grad()
#         # считаем градиент (частные производные loss-функции):
#         self.loss.backward(retain_graph=True)
#         # чуть меняем веса против градиента (кстати, можно и наоборот: будет gradient ascent):
#         self.update_parameters(lrate)
#         count = 0
        # inspect parameters:
#         for f in self.parameters():
#             if count==0:
#                 print([torch.max(f.data), torch.max(f.grad.data), self.loss])
#             count =+ 1
            
        return self.loss.detach().numpy()

net = Net()
loss = []
x = torch.tensor([10], dtype=torch.float).reshape(1,1)
y = torch.tensor([100], dtype=torch.float).reshape(1,1)
for i in range(100):
    loss.append(net.fit(x, y, lrate=0.001))
plt.plot(loss)

```

```python
import torch
from torch.autograd import Variable
import torch.optim as optim

hist = []
def linear_model(x, w):
    return torch.matmul(w, x)


w = Variable(torch.randn(1, 2), requires_grad=True)
x = torch.randn(2, 1)
y = torch.tensor([100], dtype=torch.float)
print(w, '\n', x, '\n', y)

optimizer = optim.SGD([w], lr=0.001)

# # USING ONE OF THE OPTIMIZERS (NOT NECESSARILY WITH CREATING A NEURAL NET):
# for i in range(100):
#     optimizer.zero_grad()
#     pred = linear_model(x, w)
#     loss = (pred - y) ** 2
#     loss.backward()
#     hist.append(loss)
#     optimizer.step()

# USING DIRECT TENSOR UPDATE based on its automatically calculated gradient:
for i in range(200):
    pred = linear_model(x, w)
    loss = (pred - y) ** 2
    loss.backward(retain_graph=True)
    hist.append(loss)
       
    # the gradient can be accessed in two ways:
    # w.grad.data OR torch.autograd.grad(loss, w)[0]
    w.data -= 0.01 * w.grad.data # to update a tensor, you can access in DIRECTLY via w.data.
    w.grad.data.zero_() # we zero out the gradient AFTER we do backprop

plt.plot(hist)

# now let's check if the optimized tensor (w) is indeed such that
# w.dot(x) = y

print(w.detach().numpy().dot(x.detach().numpy()))
```

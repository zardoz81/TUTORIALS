```python
import torch
from torch.autograd import Variable
import torch.optim as optim

import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

path_to_data = '/Users/RomanKoshkin/Downloads/ERP.mat'
mat_contents = sio.loadmat(path_to_data)
X = mat_contents['ERP']
z = mat_contents['z']
```

```python
# t = np.linspace(-100, 295, 80)
# erp = np.mean(X,0)
# print(erp.shape)
# plt.plot(t, erp[16,:])
```

```python
z = z[:,1]
z = z - np.mean(z)
z = z/z.std()

C_ = np.zeros([X.shape[0],32,32])
C_z_ = np.zeros([X.shape[0],32,32])
for i in range(X.shape[0]):
    C_[i,:,:] = np.cov(X[i,:,:])
    C_z_[i,:,:] = C_[i,:,:] * z[i]
C_z_ = np.mean(C_z_,0)
C_beau_ = np.mean(C_,0)
del(X)
```

```python
w = Variable(torch.randn(1, 32), requires_grad=True)
C = torch.tensor(C_, dtype=torch.float)
C_beau = torch.tensor(C_beau_, dtype=torch.float)
C_z = torch.tensor(C_z_, dtype=torch.float)
```

```python
h = []
optimizer = optim.Adam([w])

for epoch in range(1000):
    enum = (torch.mm(torch.mm(w, C_z), w.t()))**2
    denom = torch.zeros(C.shape[0])
    for i in range(C.shape[0]):
        denom[i] = (torch.mm(torch.mm(w, C[i,:,:] - C_beau), w.t()))**2
    denom = torch.mean(denom)
    f = enum/denom
    print('epoch', epoch, 'f\t', f.item())
    loss = f
    h.append(loss.item())
    loss.backward(retain_graph=False)
    ##----------- IF you WANT to use a vanilla optimizer:
    # ##     print('gradient before updating:\t', w.grad.data.sum().item())
    # ##     print('w before updating:\n', w.data)
    #     w.data += 0.5 * w.grad.data # move along the grad for grad ascent
    # ##     print('w after updating:\n', w.data)
    #     w.grad.data.zero_()
    ##----------- END BLOCK
    
    ##----------- IF you don't want to use a vanilla optimizer, but a more sophisticated builtin one:
    w.grad.data = -w.grad.data # invert sign of the gradient for gradient ASCENT
    optimizer.step()
    optimizer.zero_grad()
    ##----------- END BLOCK


sio.savemat('/Users/RomanKoshkin/Downloads/w.mat', mdict={'w': w.detach().numpy()})

fig = plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(h)
plt.subplot(1,2,2)
plt.plot(np.diff(h))
```

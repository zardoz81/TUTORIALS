# Logistic Regression
### Watch here https://www.youtube.com/watch?v=BfKanl1aSG0

```python
import numpy as np
import matplotlib.pyplot as plt

# define the logistic function (sigmoid):
def logistic(x, k, bias):
    return np.divide(np.exp(k*(x-bias)), (1+np.exp(k*(x-bias))))

def log_odds_2_p(log_odds):
    a = np.exp(log_odds).flatten()
    b = (1 + np.exp(log_odds)).flatten()
    return a/b

def log_likelihood(p, y):
    p_ = np.zeros_like(p).flatten()
    for i in range(len(y)):
        if y[i]==0:
            p_[i] = 1-p[i]
        else:
            p_[i] = p[i]
    return np.sum(np.log(p_))
```

$$ \frac{e^{k(x-b)}}{1 + e^{k(x-b)}}$$

```python
obese = [7, 8, 10, 11, 13, 14, 15, 16, 17]
n_obese = [4, 5, 6, 6.5, 7, 4.5, 5.5, 9, 8, 7]
```

```python
# plot some data that we want to classify (obese mice, not obese mice)
plt.scatter(obese, np.ones(len(obese)))
plt.scatter(n_obese, np.zeros(len(n_obese)))
plt.legend(['obese', 'not obese'])
plt.ylabel('yes/no')
plt.xlabel('weight')
```

```python
# we want to take the lotransform the 0 and 1 (probabilities) to log()
```

```python
# now let's try to fit a logistic function to the data. We don't know the parameters yet. So let's guess:
plt.scatter(obese, np.ones(len(obese)))
plt.scatter(n_obese, np.zeros(len(n_obese)))
plt.legend(['obese', 'not obese'])
plt.ylabel('yes/no')
plt.xlabel('weight')

x = np.linspace(np.min(obese + n_obese), np.max(obese + n_obese),100)
k, bias = 1, 1
plt.plot(x, logistic(x, k, bias))
plt.axhline(0.5, linewidth=0.25)
plt.axvline(bias, linewidth=0.25)

```

```python
# we see that the fit is poor. Let change the bias:
plt.scatter(obese, np.ones(len(obese)))
plt.scatter(n_obese, np.zeros(len(n_obese)))
plt.legend(['obese', 'not obese'])
plt.ylabel('yes/no')
plt.xlabel('weight')

x = np.linspace(np.min(obese + n_obese), np.max(obese + n_obese),100)
k, bias = 1, 7
plt.plot(x, logistic(x, k, bias))
plt.axhline(0.5, linewidth=0.25)
plt.axvline(bias, linewidth=0.25)
```

```python
# now the fit is better. Can we make it even better?
```

```python
# we see that the fit is poor. Let change the bias:
plt.scatter(obese, np.ones(len(obese)))
plt.scatter(n_obese, np.zeros(len(n_obese)))
plt.legend(['obese', 'not obese'])
plt.ylabel('yes/no')
plt.xlabel('weight')

x = np.linspace(np.min(obese + n_obese), np.max(obese + n_obese),100)
k, bias = 2, 7
plt.plot(x, logistic(x, k, bias))
plt.axhline(0.5, linewidth=0.25)
plt.axvline(bias, linewidth=0.25)
```

Now we log-transform the y-axis to go from odds (**why odds**? Because probabilities shoud sum to 1, and here we have a CDF that doesn't sum to 1) to $\log(odds)$. The zero odds become -Inf after this transform and ones become +Inf after this transform. Because of these infinities we can't use least squares, so we use maximum likelihood to fit a line to the data:

```python
plt.scatter(obese, 50*np.ones(len(obese)))
plt.scatter(n_obese, -50*np.ones(len(n_obese)))
plt.legend(['obese', 'not obese'])
plt.ylabel('log(odds)')
plt.xlabel('weight')

x = np.linspace(np.min(obese + n_obese), np.max(obese + n_obese),100).flatten()
k, bias = 4, -50

# we compute the log(odds). By projecting the data onto the candidate line line:
log_odds = k*np.array([obese+n_obese]) + bias

plt.scatter([obese+n_obese], log_odds)
plt.axhline(0.5, linewidth=0.25)
plt.axvline(bias, linewidth=0.25)
```

#### Then we transform the canditate log(odds) to candidate probabilities using the forumla:
$$ p = \frac{e^{\log(odds)}}{1+e^{\log(odds)}}$$
which is just a reordering of the formula for log(odds):
$$ \log\biggr(\frac{p}{1-p}\biggr)=\log(odds)$$

```python
plt.scatter(obese, np.ones(len(obese)))
plt.scatter(n_obese, np.zeros(len(n_obese)))
plt.legend(['obese', 'not obese'])
plt.ylabel('probability')
plt.xlabel('weight')

x = np.linspace(np.min(obese + n_obese), np.max(obese + n_obese),100)
y = np.array(np.ones(len(obese)).tolist() + np.zeros(len(n_obese)).tolist()).flatten()
p = log_odds_2_p(log_odds)

plt.scatter([obese+n_obese], p)
plt.axhline(0.5, linewidth=0.25)
plt.axvline(bias, linewidth=0.25)

# now we estimate the LIKELIHOOD OF DATA GIVIEN THE GREEN CURVE:  
plt.title('Log-likelihood {:.2f}'.format(log_likelihood(p, y)))
```

As we can see from the plot above and from the log-likelihood, the fit isn't great. By the way, log-likelihood means "how likely is the data given our model".
$$ \log(likelihood) = \sum_{i}{\log(p_{i})} + \sum_{j}{\log(1-p_{j})}$$
where i - are obese samples, j - are not obese.
But the parameters $k$ and $bias$ were chosen randomly. We can find the parameters by fitting a logistic regression model: 

```python
from sklearn.linear_model import LogisticRegression
X = np.array(obese + n_obese).reshape(-1, 1)
y = np.array(np.ones(len(obese)).tolist() + np.zeros(len(n_obese)).tolist()).flatten()
clf = LogisticRegression(random_state=0, solver='lbfgs').fit(X, y)

k, bias = clf.coef_, clf.intercept_
log_odds = k*np.array([obese+n_obese]) + bias
p = log_odds_2_p(log_odds)


plt.scatter(obese, np.ones(len(obese)))
plt.scatter(n_obese, np.zeros(len(n_obese)))
plt.legend(['obese', 'not obese'])
plt.ylabel('probability')
plt.xlabel('weight')

x = np.linspace(np.min(obese + n_obese), np.max(obese + n_obese),100)
p = log_odds_2_p(log_odds)

plt.scatter([obese+n_obese], p)
plt.axhline(0.5, linewidth=0.25)
plt.axvline(bias, linewidth=0.25)

plt.title('Likelihood {:.2f}'.format(log_likelihood(p,y)))
```

Now we see that the fit is much better: both the model and the log-likelihood (which is much closer to zero).

```python
clf = LogisticRegression(random_state=0, solver='lbfgs').fit(X, y)
print('Intercept\t{}\t Bias\t{}'.format((clf.coef_), clf.intercept_))
```

### We can also optimize the parameters of our logistic function (k and b) using SGD:

```python
import torch
from torch.autograd import Variable
import torch.optim as optim

k = Variable(torch.rand(1), requires_grad=True)
b = Variable(torch.rand(1), requires_grad=True)
X = torch.tensor(X, dtype=torch.float)
y = torch.tensor(y, dtype=torch.float)
optimizer = optim.Adam([k,b])
w1 = []
w2 = []
L = []
```

the loss function is
$$ y \log{\hat y} + (1-y)\log{(1-\hat{y})}$$

```python
for epoch in range(10000):
    p = torch.exp(k*(X-b))/(1 + torch.exp(k*(X-b)))
    loss = torch.mean(y * torch.log(p).flatten() + (1.0-y) * torch.log(1.0-p).flatten())
    loss.backward(retain_graph=False)
    k.grad.data = -k.grad.data
    b.grad.data = -b.grad.data
    optimizer.step()
    optimizer.zero_grad()
    L.append(loss.item())
    w1.append(k.item())
    w2.append(b.item())

plt.subplot(1,2,1)
plt.plot(L)
plt.xlabel('epoch')
plt.ylabel('loss')

plt.subplot(1,2,2)
plt.plot(w1,w2)
plt.xlabel('slopes')
plt.ylabel('biases')
plt.title('Trajectory of k and b over epochs')
```

### By the way, the number of parameters of a logistic regression model is given by $$ N*M $$ where $N$ is the number of predictors and $M$ is the number of possible outcomes:

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
print('The number of predictors is\t {}'.format(X.shape[1]))
print('The number of outcomes is\t {}'.format(len(np.unique(y))))
```

```python
clf = LogisticRegression(random_state=0, solver='lbfgs').fit(X, y)
k, bias = clf.coef_, clf.intercept_
```

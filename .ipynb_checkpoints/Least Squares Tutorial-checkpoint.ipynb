```python
import numpy as np
import matplotlib.pyplot as plt

n = 100
x = np.arange(n).flatten()

y = np.random.rand(n,1).flatten()
y = y*(x*0.03) + x*0.01
plt.scatter(x,y, c='red')
```

## Way 1

### Partial derivatives (system of equations)

<div class="warn">
    
we need to define a line 

$${ y = kx+b }$$ 

such that the error betwee each of its y-coordinate and the y-coordinate of the corresponding (lying on the same x-coordinate) data points (${y_i}$) is minimized. We will only care about absolute values of the errors, so we'll use the squares.
$${L(k,b) = \sum_{i}{(y_i - kx_i - b)^2}}$$
We need to find to differentiate (using the chain rule) this expression and set it to zero. That's where this loss (error) will be minimized.


$$\begin{cases} \frac{\partial L}{\partial k} = -2 \sum_{i}{(y_i - kx_i-b)*x_i = 0} \\ \frac{\partial L}{\partial b} = -2 \sum_{i}{(y_i - kx_i-b)*1} = 0 \end{cases}$$

Dropping the -2 and expading the brackets and the sums (for convenience), we have:

$$\begin{cases} \frac{\partial L}{\partial k} = \sum_{i}{y_ix_i} - k\sum_{i}{x_ix_i} - b\sum_{i}{x_i} = 0 \\ \frac{\partial L}{\partial b} = \sum_{i}{y_i} - k\sum_{i}{x_i}-b\sum_{i}{1} = 0 \end{cases}$$

The sigmas are in fact dot products. Let's rewrite the system of equations to get rid of the clutter:




Now solve for $k$ and for $b$


$$\begin{cases} \mathbf{y}^\top \mathbf{x} - k \mathbf{x}^\top \mathbf{x} = b\sum_{i}{x_i}\\
\sum_{i}{y_i} - k \sum_{i}{x_i} = bn \end{cases}$$

Now lets multiply both sides of the second equation in the system by $\frac{\sum_{i}{x_i}}{n}$

$$\begin{cases} \mathbf{y}^\top \mathbf{x} - k \mathbf{x}^\top \mathbf{x} = b\sum_{i}{x_i}\\
\sum_{i}{y_i}\frac{\sum_{i}{x_i}}{n} - k \sum_{i}{x_i}\frac{\sum_{i}{x_i}}{n} = bn\frac{\sum_{i}{x_i}}{n} \end{cases}$$

The right terms of both equations are now equal. We can subtract the secod from the first to get rid (temporarily) of the term containing the unknown variable $b$, and solve for $k$. 


$$ k = \frac{\mathbf{y}^\top \mathbf{x} - \sum_{i}{y_i}a}{\mathbf{x}^\top \mathbf{x} - \sum_{i}{x_i}a}$$
where $$ a = \frac{\sum_{i}{x_i}}{n}$$



Then we'll solve similarly for $b$, for which we need to get rid of the terms contaning $k$. For this we need to multiply the second equation by $\frac{\mathbf{x}^\top \mathbf{x}}{\sum_{i}{x_i}}$ and then, to cancel the terms containing $k$, subtract this second equation from the first one, and solve for $b$:

$$ b = \frac{\mathbf{y}^\top \mathbf{x} - \sum_{i}{y_i}a}{\sum_{i}{x_i} - na}$$
where $$ a = \frac{\mathbf{x}^\top \mathbf{x}}{\sum_{i}{x_i}}$$

</div>

```python
a = np.sum(x)/n
k = (np.dot(y,x) - np.sum(y)*a)/(np.dot(x,x) - np.sum(x)*a)
print('k', k)
a = np.dot(x,x)/np.sum(x)
b = (np.dot(y,x) - np.sum(y)*a)/(np.sum(x) - n*a)
print('b', b)
```

```python
y_p = k*x + b
plt.scatter(x,y, color='red')
plt.plot(x,y_p)
```

## Way 2
WE can also use SVD (PCA) to find the best fit line. Then first eigenvector is the direction of the best fit line 

```python
X = np.vstack([x,y])
U,S,V = np.linalg.svd(X)
plt.scatter(X[0,:], X[1,:])
plt.plot([0, abs(U[0,0]*100)], [0, abs(U[1,0]*100)])
plt.title('k = {:.2f} \t b = {:.2f}'.format(U[0,0], U[1,0]))
```

### But if the data is not centered, you need to compute the eigenvectors on centered data:

```python
# Let's make some data:
sc = np.array([1.2,    3.6,  2.7,   4.8,   3.2,   5.9,  6.3,  7.1,  7.5,  9])
N1 = np.array([-0.2, -0.25, -0.5, -0.45, -0.62, -0.83, -0.6, -0.9, -0.7, -1])
X = np.vstack([sc,N1])

# center the data by subtracting the mean:
b = np.mean(X,1) # the mean of the data along the y-axis
X_centered = X - b.reshape(2,-1)

# compute the eigenvectors ON CENTERED DATA:
u1, s1, v1 = np.linalg.svd(X_centered)

plt.figure(figsize=(15,5))

plt.subplot(121)
plt.scatter(X_centered[0,:], X_centered[1,:])
plt.scatter(u1[0,0], u1[1,0])

x0, y0 = 0, 0
x1, y1 = u1[0,0], u1[1,0] # largest eigenvector computed on centered data
plt.title('Centered data')
plt.plot([x0,x1],[y0,y1])
# y1 = k*x1 + b, but if we compute the slope, k, on centered data, we don't need bias because it has been subtracted
k = y1/x1
x = np.linspace(-4,4,10)
y = k*x
plt.plot(x,y) # likewise, we don't need to add bias, because the data is fully centered


# Ok, now we know k (it is the same for both centered and uncentered data) and know the bias:
plt.subplot(122)
plt.title('Uncentered data')
plt.scatter(X[0,:],X[1,:])

x = np.linspace(0,9,10)

# compute the y-values of the best fit line, also plugging back the biases:
y = k*(x-b[0])+b[1]
# the y-intercept (or bona fide bias) in this case (if you expand k*(x-b[0])+b[1]) is -k*b[0]+b[1]

plt.plot(x,y)
print('Coeffiecients using way 2: k = {:.4f} \t b = {:.4f}'.format(k, b[1]))

y = X[1,:].reshape(-1,1)
xx = np.hstack([X[0,:].reshape(-1,1), np.ones(len(X[0,:])).reshape(-1,1)])
coeff = np.linalg.inv(np.dot(xx.T, xx)).dot(xx.T).dot(y) # beta
plt.plot(x, coeff[0]*x+coeff[1])

print('Coeffiecients using way 3: k = {:.4f} \t b = {:.4f}'.format(coeff[0][0], coeff[1][0]))
```

## Way 3

See your related [question on stackexchange](https://math.stackexchange.com/questions/3205293/where-does-the-identity-matrix-come-from-in-the-formula-for-ridge-regression-coe). What we want is to minimize the **sum** of squared residuals (errors) between the predicted and real data points. The predicitons can be though of as a weighted sum (by some vector of coefficients $\beta$) of data. In the method above we projected centered data on the largest eigenvector (but lost information about the bias term, which would have to be added later). But in this method we don't lose the bias, because we do a trick and plug the intercept (bias) coefficient into the weight (coefficient) vector $\beta$ as the last element. We also add an additional 'variable' to the data as the last column of ones. (**Column, if your predictors are in columns. Don't get confused, because in EEG predictors are in rows, while time samples are in columns**)

**NOTE:**

$\mathbf X$ is the matrix of predictor variables + a column of ones

$\mathbf y$ is a column vector of true (not predicted) response variables.

$\beta$ is a column vector of coefficients (essentially slopes, whose number is equal to the number of predictors) whose last element is the bias. 

<div class="warn">
$$ L = \frac{1}{2}||\mathbf{X} \mathbf{\beta} - \mathbf{y}||^2 $$
The double bars mean norm, i.e. the **sum** of squared residuals. I.e. inside the double bars we get a vector of residuals, we square each one of them and then add them up.
    
For example, if 
$
\mathbf{X} = \begin{bmatrix}
1 & 1 \\
3 & 1 \\
0 & 1 \end{bmatrix}
$
and $ \beta =\begin{bmatrix}
1 \\
2 \end{bmatrix}
$
then the prediction (i.e. the projection of data on the coefficients, or weights) equals: $ \mathbf{X}\beta = \begin{bmatrix} 3 \\ 5 \\ 2 \end{bmatrix}$

<div class="warn">
$$ \frac{dL}{d\beta} = (\mathbf{X}\beta-\mathbf{y})^{\top}\mathbf{X} $$

set the derivative to zero, expand, and rearrange, multiply by inverse from the left (remembering to transpose the multiplier ($X$ in this case) to make sure that the inner dimensions of the matrices agree:


$$ (\mathbf{X}^{\top}\mathbf{X})^{-1}\mathbf{X}^{\top}\mathbf{X}  \mathbf{\beta} = (\mathbf{X}^{\top}\mathbf{X})^{-1} \mathbf{X}^{\top}\mathbf{y} $$

$$\mathbf{\beta} = (\mathbf{X}^{\top}\mathbf{X})^{-1} \mathbf{X}^{\top}\mathbf{y} $$

```python
l = X.shape[1]
x = X[0,:].reshape(l,1)
x = np.concatenate((x, np.ones((l,1))), axis=1)
y = X[1,:].reshape(l,1)
# beta = np.linalg.inv(np.dot(x.T,x))
beta = (np.linalg.inv(np.dot(x.T,x))).dot(x.T).dot(y)
print('k = {:.4f} \t b = {:.4f}'.format(beta[0,0], beta[1,0]))
```

## How good is the fit?

```python
x = x[:,0].flatten()
```

```python
k = beta[0,0]
b = beta[1,0]
y = k*x + b

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.scatter(X[0,:], X[1,:])
plt.scatter(x,y)
# compute and plot residuals:
resid = X[1,:] - y
plt.subplot(1,3,2)
plt.hist(resid)
plt.title('Residuals')
plt.subplot(1,3,3)
q = plt.plot(resid)
```

<div class="warn">
## Compute goodness of fit ($R^{2}$)
$$ Goodness.of.fit = \frac{variance.explained.by.the.model}{total.variance.in.the.response.var}$$




$$ \mathbf{R^{2} = \frac{Var(y_{GT}) - Var(y_{resid})}{Var(y_{GT})}}$$

```python
print('Total variance in the response variable:\t', np.var(X[1,:]))
print('unexplaided variance:\t\t\t\t', np.var(resid))
print('Explained variance (R^2):\t\t\t', (np.var(X[1,:])-np.var(resid))/np.var(X[1,:]))
print('Check if it is correct using another method:\t', np.corrcoef(X[1,:].flatten(), y)[0,1]**2)
```

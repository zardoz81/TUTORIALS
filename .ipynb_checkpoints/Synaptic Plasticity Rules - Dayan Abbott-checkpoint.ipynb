## The BCM Rule


$$
\begin{equation}
\tau_{w}\frac{d\mathbf{w}}{dt} = v\mathbf{u}(v-\theta_{v})
\end{equation}
$$


$$
\begin{equation}
\tau_{\theta}\frac{d\theta_{v}}{dt} = v^2 - \theta_{v}
\end{equation}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from utils.magic import magic
from scipy import linalg

T = 2.50

dt = 0.001
w = np.array([0.1,0.2, 0.1, 0.2, 0.1,0.1,0.1,0.05, 0.05, 0.05], dtype='float64')
# u = np.array([1,2,1,0,1,2,3,4,2,1], dtype='float64')
u = np.array([-1., -2., -1., 4., -1., -2., -3., -4., -2., -1.], dtype='float64')
r = np.random.rand(10,)

t = np.arange(0, T, dt)

tau_w, tau_theta = 2, 0.001
theta = 3
```

```python
V, THETA = [], []

for i in range(len(t)):
    u += 0.001 * r
    v = np.dot(u,w)
    V.append(v)
    dtheta = (v**2 - theta)/tau_theta*dt
    theta += dtheta
    THETA.append(theta)
    dw = v*u*(v-theta)/tau_w*dt
#     dw = 0.001*np.dot(np.outer(u,u), w)
    w += dw
    
plt.figure(figsize=(15,5))
plt.subplot(2,1,1)
plt.plot(t,V)
plt.ylabel('V')
plt.grid()
plt.axvline(t[np.argmin(np.abs(V))], color='green', linewidth=0.5)

plt.subplot(2,1,2)
plt.ylabel('theta')
plt.plot(t,THETA)
plt.grid()

plt.axvline(t[np.argmin(np.abs(V))], color='green', linewidth=0.5)


plt.tight_layout()
```

## Multiplicative Normalization and the Oja Rule


$$
\boxed{
\tau_{w}\frac{d\mathbf{w}}{dt} = v\mathbf{u} - \alpha v^2\mathbf w
}
$$


```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2)

T = 2.50

dt = 0.01
w = np.array([0.1, 0.2, 0.1, 0.2, 0.1, 0.1, 0.1, 0.05, 0.05, 0.05], dtype='float64')
u1 = np.array([-1 , -2, -1  , 4,    -1,   2,  -3,   -4,   -2,   -1], dtype='float64')
u2 = np.random.randn(10,)
v0 = 10
t = np.arange(0, T, dt)
alpha = 0.02

tau_w = 2
```

```python
V = []
D = []
u = u2
for i in range(len(t)):
    
    if t[i] > 1:
        u = u1
    if t[i] > 1.5:
        u = u2
    v = np.dot(w,u)
    V.append(v)
    dw = (v*u - alpha * v**2 * w) / tau_w * dt
    w += dw
    D.append(np.sum(np.abs(dw/dt)))

    
plt.figure(figsize=(15,5))
plt.plot(t, V)
# plt.plot(t, D)


plt.tight_layout()
```

## In this example we show that Hebbian learning forces the weight vector to align with the direction of maxumum variance of the data (i.e. the eigenvector corresponding to the biggest eigenvalue)

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2)

T = 5.50
dt = 0.01
alpha = 0.1
tau_w = 0.4 # the less tau is, the faster w grows
batch = 5

mean, cov = (0, 0), np.array([[ 1.30163034, -1.11027617], [-1.11027617,  1.32507716]])
t = np.arange(0, T, dt)
```

We'll run simulations using **four** slightly different weight update rules (Types):
$$
\begin {equation}
\tau_w \frac{d\mathbf{w}}{dt} = \mathbf{Cw}
\end {equation}
$$
where $\mathbf{C} = \langle \mathbf{uu} \rangle - {\langle \mathbf{u} \rangle}^2 = \operatorname {E} [\mathbf {u} \mathbf {u} ^{T}]- \operatorname {E}{[\mathbf{u}]}^2 $ = `cov(u)`


$$
\begin {equation}
\tau_w \frac{d\mathbf{w}}{dt} = \mathbf{Cw} - \alpha v^2 w
\end {equation}
$$

$$
\begin {equation}
\tau_w \frac{d\mathbf{w}}{dt} = v \langle \mathbf{u} \rangle - \alpha v^2 w
\end {equation}
$$

$$
\begin {equation}
\tau_w \frac{d\mathbf{w}}{dt} = v \langle \mathbf{u} \rangle
\end {equation}
$$

where $v = \langle \mathbf{u}\rangle \mathbf{w}$

(1) - **covariance rule**

(2) - **covariance rule** with Oja rule to prevent unlimited growth of the length of $\mathbf{w}$

(3) - **Basic Hebb rule** with Oja rule to prevent unlimited growth of the length of $\mathbf{w}$


```python
def run_sim(T, t, dt, alpha, tau_w, batch, cov, w, Type=1):
    U, ang, W = [], [], []
    for i in t:
        # we take a batch of input patterns (u) taken from a multivariate normal distribution:
        u = np.random.multivariate_normal(mean, cov, (batch,))
        # produce an average output evoked by the patterns in the batch:
        # critically, v is assumed to stay constant over the patterns in one batch
        v = np.mean(w.reshape(2,1).T.dot(u.T))
        C = np.cov(u.T)
        U.append(u)
        W.append(np.linalg.norm(w))

        if Type==1:
            # here the norm of the w vector is can grow without bound (Covariance Rule):
            dw = (np.dot(C,w))/tau_w*dt  # Eq. 8.10 (see eq. 1 above)
        if Type==2:
            # the norm of vector w is bounded (Covariance Rule). The more v becomes, the slower w changes.
            # Bounded Covariance Rule (Oja Rule):
            dw = (np.dot(C,w) - alpha * v**2 * w)/tau_w*dt
        if Type==3:
            dw = (v*np.mean(u, axis=0))/tau_w*dt
        if Type==4:
            dw = (v*np.mean(u, axis=0) - alpha * v**2 * w)/tau_w*dt

        w += dw
        ang.append(np.arctan(w[0]/w[1])/np.pi*180)
    u = np.vstack(U)
    U,_,_ = np.linalg.svd(u.T)

    plt.figure(figsize=(18,5))
    plt.subplot(1,3,1)
    plt.scatter(u[:,0], u[:,1])
    plt.plot([0, U[0,0]], [0, U[1,0]], linewidth=3, color='red')
    plt.plot([0, w[0]], [0, w[1]], linewidth=1, color='yellow')
    plt.legend(['1st eigenvector of the data cov m.', 'weight vector'])
    plt.grid()

    plt.subplot(1,3,2)
    plt.plot(t,ang)
    plt.ylabel('Angle of w, degrees')
    plt.grid()

    plt.subplot(1,3,3)
    plt.plot(t, W)
    plt.ylabel('${|\mathbf{w}|}^2$')
```

```python
w = np.random.rand(2,)
run_sim(T, t, dt, alpha, tau_w, batch, cov, w, Type=1)
```

```python
w = np.random.rand(2,)
run_sim(T, t, dt, alpha, tau_w, batch, cov, w, Type=2)
```

```python
w = np.random.rand(2,)
run_sim(T, t, dt, alpha, tau_w, batch, cov, w, Type=3)
```

```python
w = np.random.rand(2,)
run_sim(T, t, dt, alpha, tau_w, batch, cov, w, Type=4)
```

## Fitting the weights of synapses of two neurons. It's a network that maps a 25-D input to a 2D vector.
Here we take an image, flatten it to a 25D vector and project it on a new basis. For simplicity we don't write 25 dimensions, but only 4:

$$
\mathbf{u}^{\top} \mathbf{W} = \mathbf{v}_{pred}^{\top}
$$


$$
\begin{bmatrix}
1&2&3&4
\end{bmatrix}
\begin{bmatrix}
a & b \\
c & d \\
e & f \\
g & h
\end{bmatrix}
=
\begin{bmatrix}
a & b
\end{bmatrix}
$$

here we use the **perceptron rule** (which I figured out myself based on the Hebb rule that $\tau_w \frac{d\mathbf{w}}{dt} = \mathbf{v}_{pred} \mathbf{u}^{\top}$ and that in supervised learning we want the speed of weight change to approach zero when our predictions get closer to the targets:
$$
\boxed{
\tau_w \frac{d\mathbf{w}}{dt} = (\mathbf{v}_{pred} - \mathbf{v}_{true})\mathbf{u}^{\top}
}
$$

## one interesting way to derive the same equation is to note that we want to minimize the prediction error:

$$
E = (\mathbf{v}_{pred} - \mathbf{v}_{true})
$$
over all the samples $m$ for $m = 1,...,M$

$$
E = \sum_{m=1}^{M}(\mathbf{v}_{pred} - \mathbf{v}_{true})
$$

and to make sure the errors don't sum to zero, we take squares of the errors:

$$
E = \sum_{m=1}^{M}(\mathbf{v}_{pred} - \mathbf{v}_{true})^2
$$

We take the gradient of $E$ with respect to $\mathbf{w}$ (using the Chain rule and noting that $\mathbf{v}_{true}$ doesn't depend on $\mathbf{w}$:)

$$
\nabla_{\mathbf{w}}E = 2\sum_{m=1}^{M}(\mathbf{v}_{pred} - \mathbf{v}_{true}) \nabla_{\mathbf{w}} \mathbf{v}_{pred}
$$

we know that $\mathbf{v}_{pred} = \mathbf{u}^{\top} \mathbf{w}$ and hence $\nabla_{\mathbf{w}}\mathbf{v}_{pred} = \mathbf{u}^{\top}$. Thus, we can write:

$$
\nabla_{\mathbf{w}}E = 2\sum_{m=1}^{M}(\mathbf{v}_{pred} - \mathbf{v}_{true}) \mathbf{u}^{\top}
$$

Since we are going to set the gradient to zero, we can get rid of the factor 2 since it won't change anything:

$$
\nabla_{\mathbf{w}}E = \sum_{m=1}^{M}(\mathbf{v}_{pred} - \mathbf{v}_{true}) \mathbf{u}^{\top}
$$

We can use the **delta rule** for SGD:
$$
\mathbf{w} \rightarrow \mathbf{w} + \epsilon \nabla_{\mathbf{w}} E
$$

```python
loss = []
u = np.zeros((5,5))
u[2,:] = 1
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.imshow(u)
W = np.random.randn(u.flatten().shape[0],2)
v_ = np.array([0.1, 0.9]).reshape(2,1)

print(np.dot(W.T, u.reshape(-1,1)))

for i in range(100):
#     dW = (v.dot(u.reshape(1,-1)) - alpha * W.T * v**2)/tau_w*dt
    v = np.dot(W.T, u.reshape(-1,1))
    dW = (np.dot(v_ - v, u.reshape(1,-1)) - alpha * W.T * v**2)/tau_w*dt
    loss.append(np.linalg.norm(v_ - v))

    W += dW.T

print(np.dot(W.T, u.reshape(-1,1)))
plt.subplot(1,2,2)
plt.plot(loss)
```

## In this example we feed two different images and their corresponding labels:

```python
loss = []
u1 = np.zeros((5,5))
u1[2,:] = 1
u2 = np.zeros((5,5))
u2[:,2] = 1
u = [u1, u2]

W = np.random.randn(u[0].flatten().shape[0],2)
v_1 = np.array([0.1, 0.9]).reshape(2,1)
v_2 = np.array([0.9, 0.1]).reshape(2,1)
v_ = [v_1, v_2]

plt.figure(figsize=(15,5))
plt.subplot(1,4,1)
plt.title('Label:\n'+ str(v_[0]))
plt.imshow(u1)
plt.subplot(1,4,2)
plt.imshow(u2)
plt.title('Label:\n'+ str(v_[1]))

print('Prediction before training')
print(np.dot(W.T, u[0].reshape(-1,1)))
print(np.dot(W.T, u[1].reshape(-1,1)))

for i in range(300):
    v = np.dot(W.T, u[i%2].reshape(-1,1))
    dW = (np.dot(v_[i%2] - v, u[i%2].reshape(1,-1)) - alpha * W.T)/tau_w*dt
#     dW = (np.dot(v_[i%2] - v, u[i%2].reshape(1,-1)))/tau_w*dt
    loss.append(np.linalg.norm(v_[i%2] - v))
    W += dW.T

print('Prediction after training')
print(np.round(np.dot(W.T, u[0].reshape(-1,1)), 2))
print(np.round(np.dot(W.T, u[1].reshape(-1,1)), 2))
plt.subplot(1,4,(3,4))
plt.plot(loss)
plt.ylabel('Loss')
```

## Function approximation
### The neurons in the input layer have tuning curves that we don't fit, just set. We only fit the weights.

```python
class FuncApprox():
    def __init__(self):
        self.epsilon = 0.01
        self.W = np.random.randn(11,1)
        
    def f(self, s):
        # returns the value of a basis function in a given x-point:
        u = []
        for mu in np.linspace(-10,10,11):
            u.append(np.exp(-0.5*(s-mu)**2))
        return np.vstack(u)

    def fit(self, I):
        loss = []
        for i in range(I):
            s = np.random.choice(np.linspace(-10,10,100))
            u = self.f(s)
            v_true = np.sin(s)
            v_pred = u.T.dot(self.W)
            self.W -= self.epsilon * np.dot((v_pred-v_true).T, u.T).T
            loss.append(((v_pred- v_true)**2).flatten())
        return loss

    def predict(self, x):
        u = self.f(x)
        return u.T.dot(self.W)

fa = FuncApprox()
x = np.linspace(-10,10,1000)
loss = fa.fit(10000)
pred = fa.predict(x)

plt.figure(figsize=(15,5))
plt.plot(x, np.sin(x))
plt.plot(x, pred)
plt.legend(['Ground Truth', 'Predicted'])

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(loss)
plt.title('Loss')

plt.subplot(1,2,2)
for mu in np.linspace(-10,10,11):
    plt.plot(x, np.exp(-0.5*(x-mu)**2))
plt.title('Basis functions')
```

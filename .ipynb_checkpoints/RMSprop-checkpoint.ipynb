```python
import numpy as np
import matplotlib.pyplot as plt

beta = 0.1
epsilon = 1e-8

n = 50
x = np.linspace (0.001, 2*np.pi, n)
y = np.sin(x) + np.random.randn(n,)
dw = np.append(np.array([0]), np.diff(y))

sdw = [0]
y_ = [0]
for dw_ in dw[:-1]:
    sdw.append(sdw[-1] * beta + (1-beta) * dw_**2)
    y_.append(dw_/(np.sqrt(sdw[-1]) + epsilon))
```

```python
plt.figure(figsize=(15,5))
plt.plot(x, y)
plt.stem(x, dw, use_line_collection=True)

yy = dw/(np.sqrt(sdw) + epsilon)
yy[0:2] = 0
plt.plot(x, yy)
```

The simple integrate-and-fire equation is:
$$
\tau_m \frac{dV}{dt} = E_L - V + R_mI_e
$$

We need to find a solution to it.

First, we move the $\tau_m$ to the other side of the equation:

$$
\frac{dV}{dt} = \frac{E_L}{\tau_m} - \frac{V}{\tau_m } + \frac{R_mI_e}{\tau_m }
$$

Then we need to separate the variables. This means that all the V-containing terms will be one one side, and the other variable, $t$, to the other. This is crucial (refer to the pdf file "Solving differential equations"):

$$
\frac{dV}{dt} = \frac{E_L - V + R_mI_e}{\tau_m }
$$

In any proportion we can flip the elements diagonally. For example the numerator of the left term with the denomination of the right term. By the way we can also take the reciprocal of both terms (at the same time, of course). So let's use this trick to separate the variables:

$$
\frac{dV}{E_L - V + R_mI_e} = \frac{dt}{\tau_m }
$$

We'll now integrate both sides. To make things clearer, let's rewrite the terms a little:

$$
\frac{1}{E_L - V + R_mI_e}dV = \frac{1}{\tau_m }dt
$$

Now integrate:

$$
\int \frac{1}{E_L - V + R_mI_e}dV = \int \frac{1}{\tau_m }dt
$$

$$
-\ln \bigg|E_L - V + R_mI_e\bigg| = \frac{t}{\tau_m } + \tilde C
$$

Every integral come with a constant. Here we write one $\tilde C$ because we put the two constants together.

Now let's move the minus to the right side and get rid of the natural log on the left by raising e to the power of both sides:

$$
\exp \bigg(\ln \bigg|E_L - V + R_mI_e\bigg|\bigg) = \exp \bigg(-\big(\frac{t}{\tau_m} + \tilde C\big) \bigg)
$$

$$
E_L - V + R_mI_e = \exp \big(-\frac{t}{\tau_m}\big) \exp \big(-\tilde C\big)
$$

Let $\exp(-\tilde C) = C$:

$$
E_L - V + R_mI_e = Ce^{-\frac{t}{\tau_m}}
$$

and

$$
V(t) = E_L + R_mI_e - Ce^{-\frac{t}{\tau_m}}
$$

If at $t = 0$ our neuron's membrane potential is zero (i.e. $V(0) = 0$), then $C$ must be equal to $E_L+R_mI_e$ (because that's the only way to get V to be zero at $t=0$). 

$$
V(t) = E_L + R_mI_e - (E_L + R_mI_e)e^{-\frac{t}{\tau_m}}
$$


One the other hand, at $t=0$ our neuron can already have some potential. To accommodate this possibility, we need to add to our equation the initial condition term, $V(0)$:

$$
\boxed{
V(t) = E_L + R_mI_e - (E_L + V(0) + R_mI_e)e^{-\frac{t}{\tau_m}}
}
$$

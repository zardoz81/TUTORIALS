```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = 100
fs = 2

gauss_ker = np.exp(-(np.linspace(-4,4,30*fs))**2)

u = np.zeros((T*fs,))
u[20*fs] = 1
u = np.convolve(u, np.hstack([np.linspace(0,1,2*fs), np.linspace(1,0,2*fs)]), mode='same')
w = np.zeros((T*fs,))

r = np.zeros((T*fs,))
r[70*fs] = 1

r = np.convolve(r, gauss_ker, mode='same')

def get_v(t, w, u):
    v = []
    for tau in range(t):
        v.append(w[tau]*u[t-tau])
    return np.sum(v)

def get_delta(r, v1, v0):
    return r+v1-v0

def fit(u, w, epsilon, delta, t):
    for tau in range(t):
        w[tau] = w[tau] + epsilon*delta*u[t-tau]
    return w
    

def run_one_trial(T,u,w,r,fs):
    # compute prediction of total future reward at each time step:
    v = []
    for t in range(T*fs):
        v.append(get_v(t,w,u))
    v = np.array(v)

    # compute temporal difference between predictions of total future reward at each time step:
    delta_v = []
    for t in range(0, T*fs-1):
        delta_v.append(v[t+1] - v[t])
    delta_v.insert(0,0)
    delta_v = np.array(delta_v)

    # compute errors between predictions of predicted current reward and actual current reward at each time step. 
    # Note that predicted CURRENT reward equals predicted TOTAL FUTURE reward at t+1 minus total predicted future 
    # reward at t. That's delta_v
    delta = []
    for t in range(0, T*fs-1):
        delta.append(get_delta(r[t], v[t+1], v[t])) ################ here was an ERROR $$$$$$$$$$$$$$@@@@@@@@@
#         w = fit(u,w,0.01,delta[-1],t)
    delta.append(0)
    delta = np.array(delta)
    
    return w, v, delta_v, delta    

def plot_status(v, delta_v, delta, u, r, fs):
    tt = np.linspace(0,T,len(v))
    plt.figure(figsize=(15,8))
    plt.subplot(5,1,1)
    plt.plot(tt, u)
    plt.ylabel('$\mathbf{u}$', fontsize=20)
    plt.subplot(5,1,2)
    plt.plot(tt, r)
    plt.ylabel('$\mathbf{r}$', fontsize=20)
    plt.subplot(5,1,3)
    plt.plot(tt, v)
    plt.ylabel('$\mathbf{v}$', fontsize=20)
    plt.subplot(5,1,4)
    plt.plot(tt, delta_v)
    plt.ylabel('$\mathbf{\Delta V}$', fontsize=20)
    plt.subplot(5,1,5)
    plt.plot(tt, delta)
    plt.ylabel('$\mathbf{\delta}$', fontsize=20)
    plt.tight_layout()

w, v, delta_v, delta = run_one_trial(T,u,w,r, fs)
plot_status(v, delta_v, delta, u, r, fs)
```

```python
for j in range(300):
    w, v, delta_v, delta = run_one_trial(T,u,w,r, fs)
    for t in range(T*fs):
        w = fit(u,w,0.01,delta[t],t)
plot_status(v, delta_v, delta, u, r, fs)
```

```python
# import pixiedust

from sklearn.metrics import auc
from skimage.util.shape import view_as_windows
import scipy.stats as ss
import nest
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from sklearn.preprocessing import normalize
from scipy.signal import butter, filtfilt
from sklearn.preprocessing import minmax_scale

from matplotlib.pylab import *
from mpl_toolkits.axes_grid1 import host_subplot
import matplotlib.animation as animation
from utils.printProgressBar import printProgressBar
from time import time
import pickle
from scipy.signal import convolve2d as conv2d
from skimage.color import rgb2gray

def fancy_kernel(t):
    tau_j = 3 # roll-off
    tau_f = 0.1  # roll-on
    V0 = 2
    t_j = 0
    return np.heaviside(t-t_j, 1) * V0*(np.exp(-(t-t_j)/tau_j) - np.exp(-(t-t_j)/tau_f))

def normal_kernel(t):
    ker_times = t[t>=0]
    tau0 = 1.7
    sign = np.ones_like(ker_times)
    tau = np.ones_like(ker_times)*tau0
    ker = sign*np.exp(-ker_times/tau)
    ker = np.concatenate([np.zeros_like(t[t<0]), ker])
    ker = ker/ker.max()
    return ker

def error_ker(times):
    tau0 = 2.8
    tau1 = 1.1
    sign = np.ones_like(times)
    tau = np.ones_like(times)*tau0
    tau[times>0] = tau1
    sign[times>0] = -1
    return sign*np.exp(sign*times/tau)

# get a desired spike train (Sd)
def get_Sd(n_spikes_in_Sd, times):    
#     Sd_times = np.round(sorted(np.random.uniform(10,90, (n_spikes_in_Sd,))), roundingfactor)
    Sd_times = [30., 40., 60., 70.]
    print('Sd_times: {}'.format(Sd_times))
    return signal.unit_impulse(len(times), np.searchsorted(times, Sd_times))

def get_delta_w(ad, E, So, Sd):
    dw = []
    for i in range(len(E)):
        dw.append((Sd-So)*(ad + E[i]))
    return np.vstack(dw)

def update_w(weights, dw):
    weights = weights + np.cumsum(deltas, axis=1)[:,-1]
#     weights[weights<0] = 0 # allow negative weights
    return weights

def get_eligibility(times, kernel, n_spike_generators, spikegenerators):
    E = []    
    for i in range(n_spike_generators):
        in_sp_times = nest.GetStatus([spikegenerators[i]])[0]['spike_times']
        st = signal.unit_impulse(len(times), np.searchsorted(times,in_sp_times))
        eligibility = np.convolve(st, kernel, mode='same')
        E.append(eligibility)
    return E

def w_sinc(Fc_Hz, dt_ms, len_ker_ms):
    len_ker_s = len_ker_ms/1000
    dt_s = dt_ms/1000
    srate = 1/dt_s    #  srate per second (not per millisecond, as in Nest)
    Fc = Fc_Hz/srate   # as fraction of the srate (but no more than 0.5, due to Nyquist)
    M = int(len_ker_s/dt_s) # length of kernel in samples
    K = 1
    h = np.zeros((M, ))
    for i in range(M):
        if i == M/2:
            h[i] = 2*np.pi*Fc*K
        else:
            h[i] = K * np.sin(2*np.pi*Fc * (i-M/2))/(i-M/2) * (0.42 - 0.5*np.cos((2*np.pi*i)/M) + 0.08 * np.cos((4*np.pi*i)/M))
    return h

def check_n_fix_spike_times(So_times, times):
    if So_times.size > 0 and  So_times[-1] > times[-1]:
        print('fixing the trailing spike')
        So_times[-1] -= times[-1]
    return So_times

def get_C(Sd, So, B, A):
    out1 = filtfilt(B, A, So)
    out2 = filtfilt(B, A, Sd)
    C = out1.dot(out2)/(np.linalg.norm(out1) * np.linalg.norm(out2))
    return C, out1, out2

def simulate(T, dt, neuron_type, n, n_spike_generators, n_spikes, weights, in_spike_train):
    nest.ResetKernel()
    nest.SetKernelStatus({"resolution": dt})
    
    iaf_neuron_parms = {
     'C_m': 1000.0, # pF
     'E_L': -65.0, # resting potential
     'I_e': 1000.0,
     't_ref': 5.0,
     'tau_m': 10.0,
     'tau_syn_ex': 1.0,
     'tau_syn_in': 2.0,
     'V_m': -60.0, # initial potential?
     'V_reset': -65.0,
     'V_th': -55.0}

    neurons = nest.Create(neuron_type, n)   
    spikedetectors = nest.Create('spike_detector', n)
    spikegenerators = nest.Create('spike_generator', n_spike_generators)
    voltmeters = nest.Create('voltmeter', n, params={"interval": dt})
    multimeter = nest.Create('multimeter', 1,  params = {
        'withtime':True,
        'interval':dt,
        'record_from':['V_m', 'I_syn_ex', 'I_syn_in']})

    nest.Connect(spikegenerators, neurons, 'all_to_all', syn_spec={'weight': weights})
    nest.Connect(multimeter, neurons, 'one_to_one')
    nest.Connect(voltmeters, neurons, 'one_to_one')
    nest.Connect(neurons, spikedetectors, 'one_to_one')
    
    if neuron_type=='iaf_psc_alpha':
        nest.SetStatus(neurons, params=iaf_neuron_parms)
    c = 0
    for i in spikegenerators:
        nest.SetStatus([i], {'spike_times': in_spike_train[c]})
        c += 1
#     nest.SetStatus([neurons[0]], {'I_e':0.0})
    nest.Simulate(T+dt)
    
    times = nest.GetStatus(voltmeters)[0]['events']['times']
    return times, voltmeters, spikedetectors, multimeter, spikegenerators


def gabor(sigma=1, theta=0, Lambda=1, psi=0, gamma=1):
    """
    Gabor feature extraction.
    sigma - length of a standard deviation (pixels)
    lambda - wavelength
    psi - phase shift
    gamma - ellipticity
    
    Usage:
    gabor(sigma=5, theta=np.pi/4, Lambda=6, psi=0, gamma=1)
    will give you a round (not elliptical) gabor filter of size (5,5) and a wavelength of 6
    """
    sigma_x = sigma
    sigma_y = float(sigma) / gamma

    # Bounding box
    nstds = 1  # Number of standard deviation sigma
    xmax = max(abs(nstds * sigma_x * np.cos(theta)), abs(nstds * sigma_y * np.sin(theta)))
    xmax = np.ceil(max(1, xmax))
    ymax = max(abs(nstds * sigma_x * np.sin(theta)), abs(nstds * sigma_y * np.cos(theta)))
    ymax = np.ceil(max(1, ymax))
    xmin = -xmax
    ymin = -ymax
    (y, x) = np.meshgrid(np.arange(ymin, ymax + 1), np.arange(xmin, xmax + 1))

    # Rotation
    x_theta = x * np.cos(theta) + y * np.sin(theta)
    y_theta = -x * np.sin(theta) + y * np.cos(theta)

    gb = np.exp(-.5 * (x_theta ** 2 / sigma_x ** 2 + y_theta ** 2 / sigma_y ** 2)) * np.cos(2 * np.pi / Lambda * x_theta + psi)
    return gb
```

```python
# import pixiedust

from skimage.util.shape import view_as_windows
import scipy.stats as ss
import copy

from skimage.util.shape import view_as_windows
import scipy.stats as ss
import time
import pickle
import numpy as np
import matplotlib.pyplot as plt
from time import time
from scipy import signal
from scipy.stats import norm
from scipy.special import softmax
from scipy.ndimage.filters import gaussian_filter1d as gauss
from sklearn.metrics import auc
from sklearn.preprocessing import minmax_scale
from sklearn.datasets import fetch_openml
from scipy.signal import argrelextrema
from scipy import signal
import skimage
from skimage.measure import block_reduce as MaxPooling2D
from skimage.transform import resize as imresize
from utils.printProgressBar import printProgressBar
if not 'mnist' in locals():
    mnist = fetch_openml('mnist_784', version=1, cache=True)

def get_maxima(freq, t_max, dt, in_dim):

    srate = np.round(1.0/dt, 5)
    period = np.floor(srate/freq) # period
    print('Full period {:.0f} samples \nWhich means that even if all the pixels have the same value they will still get properly encoded by their unique spike timings'.format(period))
    phase_step = int(period/in_dim)
    print('Number of pixels in the image: {}'.format(in_dim))
    print('Phase step {:.0f} samples'.format(phase_step))
    
    Ts = int(srate/freq)
#     print('This sampling rate and sin freq can accommodate {} uniquely times spikes'.format(Ts))
    t = np.arange(0, t_max, dt)
    y = np.cos(2*np.pi*freq*t)
    maxima = np.nonzero(y==y.max())[0]
    TTT = []
    tt = np.nonzero(y==1.0)[0]
    for i in range(Ts):
        ttt = np.arange(maxima[0]-phase_step*i, len(t), period)
        if ttt.min()<0:
            ttt = ttt[ttt>=0]
        TTT.append(t[ttt.astype('int')])
    return TTT, y
    

def align(spike_timings, maxima):
#     print('DT in the align func', dt)
    t = np.round(np.arange(0, t_max, dt), 5) # rounding factor !!!
    SST = []
    for i in range(len(spike_timings)):
        snapped_to_peak_spike = np.argmin(np.abs(maxima[i] - spike_timings[i]))
        snapped_latency = maxima[i][snapped_to_peak_spike]
        SST.append(np.round(snapped_latency, 5))
    return np.array(SST)

def get_sp_latency(p_intensities, t_max=1, p_intensity_min=1, p_intensity_max=256):
    t_max_fact = t_max/(np.log(p_intensity_max) - np.log(p_intensity_min))
    spike_timings = t_max_fact * (np.log(p_intensity_max) - np.log(p_intensities))
    return p_intensities, spike_timings

def get_ganglion(SST):
    """
    this function gets ganglion cells' receptive fields
    """
    gcrf = 4                                # size of a ganglion cell's square RF in pixels
    x = int(np.sqrt(SST.shape[0]/gcrf**2))  # number of RFs along each dimension (H and W)
        
    ganglion_input = np.zeros((x,x,gcrf,gcrf))
    for i in range(x):
        for j in range(x):
            ganglion_input[i,j,:,:] = SST.reshape(imsize)[i*4:4+i*4, j*4:4+j*4]
    return ganglion_input, x
```

### then we make five scaled versions of the image (100, 71, 50, 35, 25 percent of the original number of pixels:

```python
# %%pixie_debugger


def get_scaled(im, imsize, pad=False):
    IM = []
    im = rgb2gray(im)
    im = imresize(im, imsize)
    org_s = im.shape[0]
    percentages = [1, 0.75, 0.5, 0.35, 0.25]
    for p in percentages:
        s = int(np.sqrt(im.flatten().shape[0]*p)) #  
        im_ = imresize(im, (s, s))
        if pad:
            pad_with = int(np.floor((org_s-s)/2)) + 1
            im_ = np.pad(im_, pad_width=pad_with, mode='constant', constant_values=0)
            print(im_.shape)
            im_ = im_[:imsize[0], :imsize[1]]
        IM.append(im_)
    return IM

def get_filters(THETAS):
    THETAS = [np.pi/8, pi/4 + np.pi/8, np.pi/2 + np.pi/8, 3*np.pi/4 + np.pi/8]
    sigma = 2
    Lambda = 5
    gamma = 1
    G = []
    for theta in THETAS:
        G.append(gabor(sigma=sigma, theta=theta, Lambda=Lambda, psi=0, gamma=gamma))
#         plt.subplot(1,4, i+1)
#         plt.imshow(G[-1])
    return G

def get_S1_maps(scaled_IM, kernels, input_size):
    S1maps = {}
    for im in range(len(scaled_IM)): # loop over the scaled versions of the image
        S1maps['scale'+str(im)] = np.zeros((4,) + scaled_IM[im].shape)
        for k in range(len(kernels)):# loop over the filters
            S1map = minmax_scale(conv2d(scaled_IM[im], kernels[k], mode='same').flatten(), (0, 255-0.01)).reshape(scaled_IM[im].shape)
            S1maps['scale'+str(im)][k,:,:] = S1map
    return S1maps

def get_C1_maps(S1maps):
    C1maps = {}
    for scale in S1maps.keys():
        C1maps[scale] = []    # init empty list for one processing scale
        for orientation in range(S1maps[scale].shape[0]):
            # break into receptive fields (windows of size 7 with stride 6)
            X = view_as_windows(S1maps[scale][orientation,:,:], window_shape = (7,7), step=(6,6))
            C1map = np.zeros((X.shape[0], X.shape[1]))
            for i in range(X.shape[0]):
                for j in range(X.shape[1]):
                    X_r = ss.rankdata(X[i,j,:,:].flatten(), method='ordinal').reshape(7,7)
                    C1map[i,j] = X[(i,j) + np.where(X_r==X_r.max())]
            C1maps[scale].append(C1map) # append to the current processing scale another orientation map
        C1maps[scale] = np.stack(C1maps[scale], axis=0)
    return C1maps



input_size = (128, 128)
im = skimage.data.coffee()

G = get_filters(THETAS)           # get the FOUR spatial filters
IM = get_scaled(im, input_size, pad=False)   # returns a list of scaled version of the input image
S1maps = get_S1_maps(IM, G, input_size) # returns an array of size (n_ker, n_scales, H, W)

fig = plt.figure(figsize=(12,8))  # (width, height, NOT vice versa)
c = 0
for i in range(4):
    ax = plt.subplot(4, 6, c+1)
    ax.imshow(G[i], cmap='gray')
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    c += 1
    ax.set_axis_off()
    for scale in S1maps.keys():
        ax = plt.subplot(4, 6, c+1)
        ax.imshow(S1maps[scale][i], cmap='gray')
        ax.set_yticklabels([])
        ax.set_xticklabels([])
        c += 1
        ax.set_axis_off()
fig.subplots_adjust(top = 1, bottom = 0, right = 1, left = 0, wspace=0, hspace=0)
plt.suptitle('S1 maps', weight='bold', c='yellow', size=25)
```

```python
# then we perform lateral inhibition of spike latencies in the C1 layer. I.e. the earliest
# spikes delay nearby spikes using the below Gaussian-like mask. The inhibition is strongest
# for the spikes that happen to be near and fades out to zero to spikes located 6 pixels away from the current one

def inv_gauss_mask(size, fwhm=3, center=None):
    """ Make a square gaussian kernel. 
    size is the length of a side of the square 
    fwhm is full-width-half-maximum, which can be thought of as an effective radius."""
    x = np.arange(0, size, 1, float)
    y = x[:,np.newaxis]
    if center is None:
        x0 = y0 = size // 2
    else:
        x0 = center[0]
        y0 = center[1]
    g = np.exp(-4*np.log(2) * ((x-x0)**2 + (y-y0)**2) / fwhm**2) + 1
    g[x0, y0] = 1
    return g

plt.imshow(inv_gauss_mask(11, fwhm = 6, center=None))
plt.colorbar()
```

```python
import copy

def encode_C1maps(C1maps, snap=2): 
    """ covert pixel intensities to lantecies """
    C1maps_ = copy.deepcopy(C1maps)
    for k in C1maps_.keys():
        for j in range(C1maps_[k].shape[0]):
            C1maps_[k][j] = np.abs(np.round(255 - C1maps_[k][j], snap)) # snap to 0.01 ms steps
    return C1maps_

def inhibit_laterals(input_im, out, span=5, min_att=1.05, max_att=1.15, t=-1):
    X = np.copy(input_im)
    if t >= 0:
        I, J = np.where(X==t)
        if I.size == 0:
            I, J = np.where(X==np.min(X))
            return X, I[0], J[0], np.min(X[X>t])
    if t==-1:
        I, J = np.where(X==np.amin(X))
    factor = np.linspace(min_att, max_att, span).tolist() + [1.0] + np.linspace(max_att, min_att, span).tolist()
    
    g = inv_gauss_mask(11, fwhm = 6, center=None)

    for c0,i in enumerate(np.arange(I[0]-span, I[0]+span+1)):
        for c1,j in enumerate(np.arange(J[0]-span, J[0]+span+1)):
            if i>=0 and i<X.shape[0] and j>=0 and j<X.shape[1]:
                f = g[c0,c1]
                if X[i,j] >= t:
                    X[i,j] = np.round(X[i,j]*f, 2)
            else:
                continue
    next_spike_t = np.min(X[X>t])
    return X, J[0], I[0], next_spike_t

C1maps_encoded = encode_C1maps(C1maps, snap=2)

        
X = C1maps_encoded['scale0'][1]
X_, i, j, next_spike_t  = inhibit_laterals(X, out, span=5, min_att=1.05, max_att=1.15, t=-1)

plt.figure(figsize=(12,5))
plt.subplot(1,3,1)
plt.imshow(C1maps['scale0'][0], vmin=0, vmax=255)
plt.colorbar(orientation="horizontal")
plt.scatter(i, j, c='red', s=40) # beware! Scatter takes (x,y), while imshow takes (y, x)
plt.title('C1 maps before converting to latencies')
plt.subplot(1,3,2)
plt.imshow(X, vmin=0, vmax=255)
plt.colorbar(orientation="horizontal")
plt.scatter(i, j, c='red', s=40) # beware! Scatter takes (x,y), while imshow takes (y, x)
plt.title('C1 maps after converting to latencies')
plt.subplot(1,3,3)
plt.imshow(X_, vmin=0, vmax=255)
plt.colorbar(orientation="horizontal")
plt.scatter(i, j, c='red', s=40) # beware! Scatter takes (x,y), while imshow takes (y, x)
plt.title('C1 maps after converting to latencies\n and inhibition (one step)')
```

```python
def new_simulate(T, dt, neuron_type, n, n_spike_generators, weights, in_spike_train):
    """implements online STDP. I don't exactly know now it works, but it does something"""
    nest.ResetKernel()
    nest.SetKernelStatus({"resolution": dt})
    
    iaf_neuron_parms = {
     'C_m': 1000.0, # pF
     'E_L': -65.0, # resting potential
     'I_e': 1000.0,
     't_ref': 5.0,
     'tau_m': 10.0,
     'tau_syn_ex': 1.0,
     'tau_syn_in': 2.0,
     'V_m': -60.0, # initial potential?
     'V_reset': -65.0,
     'V_th': -55.0}
    
    neurons = nest.Create(neuron_type, n)   
    spikedetectors = nest.Create('spike_detector', n)
    spikegenerators = nest.Create('spike_generator', n_spike_generators)
    voltmeters = nest.Create('voltmeter', n, params={"interval": dt})
    multimeter = nest.Create('multimeter', 1,  params = {
        'withtime':True,
        'interval':dt,
        'record_from':['V_m', 'I_syn_ex', 'I_syn_in']})
    pre_pop = nest.Create('parrot_neuron', n_spike_generators)
    
    nest.Connect(spikegenerators, pre_pop, "one_to_one")
    nest.Connect(pre_pop, neurons,
                 conn_spec='all_to_all',
                 syn_spec={'model': 'static_synapse',
                           'weight': weights,
                           'delay': 0.01}
                )
    nest.Connect(multimeter, neurons, 'one_to_one')
    nest.Connect(voltmeters, neurons, 'one_to_one')
    nest.Connect(neurons, spikedetectors, 'one_to_one')
    
    if neuron_type=='iaf_psc_alpha':
        nest.SetStatus(neurons, params=iaf_neuron_parms)
    c = 0

    
    for i in spikegenerators:
        nest.SetStatus([i], {'spike_times': np.array(in_spike_train[c]).flatten()})
        c += 1
#     nest.SetStatus([neurons[0]], {'I_e':0.0})

    nest.Simulate(T+dt)
    weights_ = np.array(nest.GetStatus(nest.GetConnections(pre_pop, neurons), keys='weight')).reshape(1,-1)
    times = nest.GetStatus(voltmeters)[0]['events']['times']
    return times, voltmeters, spikedetectors, multimeter, spikegenerators, neurons, pre_pop, weights_


```

```python
def init_weights():
    W = {}
    for scale in C1maps_inh.keys():
        W[scale] = np.random.rand((*C1maps_inh[scale].shape))*1
#         W[scale] = np.ones_like(C1maps_inh[scale]) * 450
    return W

def get_winning_scale(out_ST):
    ealiest_spikes_in_scales = [np.min(i) for i in out_ST.values()]
    winning_scale_id = np.argmin(ealiest_spikes_in_scales)
    winning_scale_name = list(out_ST.keys())[winning_scale_id] # where the earliest spike is
    winner_spike = np.round(np.min(out_ST[winning_scale_name]), 2)
    return winning_scale_name, winner_spike

W = init_weights()
```

```python
def STDP(W, out_ST, C1maps_inh):
    ws_t = get_winning_scale(out_ST)[1] # latency of the earliest spike
    DW = {}
    for scale in C1maps_inh.keys():
        t_j = C1maps_inh[scale] # presynaptic spike times
        t_i = out_ST[scale]
        diff = t_j - ws_t # get who late or early the presynaptic spikes are relative to the winner postsynaptic spike
        a = np.zeros_like(diff) # make a vector of learning rates 
        a_plus = 0.01
        a_minus = -0.75 * a_plus
        a[diff<=0] = a_plus # (posive for those presyn spikes that preceed (i.e. cause) the winner)
        a[diff>0] = a_minus # (negative for those presyn spikes that trail (i.e. not cause) the winner)
        dw = a * W[scale] * (1-W[scale])
        W[scale] += dw
        DW[scale] = np.copy(dw)
    return W, DW

plt.ioff()
for i in range(25):
    W, DW = STDP(W, out_ST, C1maps_inh)
    fig = plt.figure(figsize=(18,3))

    plt.subplot(1,4,1)
    plt.imshow(C1maps_inh['scale0'][0,:,:])
    plt.colorbar()
    plt.title('C1 map')

    plt.subplot(1,4,2)
    plt.imshow(DW['scale0'][0,:,:])
    plt.colorbar()
    plt.title('dw')

    plt.subplot(1,4,3)
    plt.imshow(W['scale0'][0,:,:])
    plt.colorbar()
    plt.title('W')
    
    if i%5==0:
        plt.savefig('{}{}'.format('frame_', i), dpi=100, transparent=False, pad_inches=0.1)
    plt.close(fig)
```

An ODE 

describes how a variable decays exponentially towards a certain value:
$$
\tau \frac{dV}{dt} = -V
$$
or grows exponentially away from it

$$
\tau \frac{dV}{dt} = V
$$

:

```python
import numpy as np
import matplotlib.pyplot as plt

T = 10
tau = 1
dt = 0.01

t = np.arange(0, T, dt)
v = -10
dv = 0

V = []
for i in range(len(t)):
    dv = (-v)/tau*dt
    v = v + dv
    V.append(v)

t = t.flatten()
plt.plot(t,V)

```

```python
import numpy as np
import matplotlib.pyplot as plt

T = 10
tau = 1
dt = 0.01

t = np.arange(0, T, dt)
v = -10
dv = 0

V = []
for i in range(len(t)):
    dv = (v)/tau*dt
    v = v + dv
    V.append(v)

t = t.flatten()
plt.plot(t,V)

```

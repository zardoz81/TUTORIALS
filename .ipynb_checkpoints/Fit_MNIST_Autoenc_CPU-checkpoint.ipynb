```python
import numpy as np
import scipy
import pickle
# import hiddenlayer as hl

from keras.datasets import mnist
from keras.utils import to_categorical

from skimage import transform as im_tf

import matplotlib.pyplot as plt
%matplotlib inline

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.autograd import Variable

# make sure you don't hog all the video memory
import os

from sklearn.datasets import fetch_openml

from utils.printProgressBar import printProgressBar

import copy
import time

```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()

x = torch.tensor(x_train, dtype=torch.float).view(-1,1,784)

model = nn.Sequential(
    nn.Linear(784,10, bias=False),
    nn.Linear(10, 784))

optimizer = optim.Adam(model.parameters())
criterion = nn.MSELoss()

model

```

```python
NITER = 150


hist = []
for i in range(NITER):
    optimizer.zero_grad()
    predictions = model(x)
    targets = x

    loss = criterion(predictions, targets)
            
    loss.backward()
    hist.append(loss)
    optimizer.step()
    suffix = 'Loss: {:.2f}'.format(loss.item())
    printProgressBar(i, NITER-1, prefix = 'Progress:', suffix=suffix, length=50)
```

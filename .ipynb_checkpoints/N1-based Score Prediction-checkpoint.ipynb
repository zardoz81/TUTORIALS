## Suppose we get the following average data from reference ads:

```python
import numpy as np
import matplotlib.pyplot as plt


def fit(X):
    y = X[1,:].reshape(-1,1)
    xx = np.hstack([X[0,:].reshape(-1,1), np.ones(len(X[0,:])).reshape(-1,1)])
    coeff = np.linalg.inv(np.dot(xx.T, xx)).dot(xx.T).dot(y) # beta
    return coeff[0], coeff[1]
    
def permute(X):
    y = X[1,:].reshape(-1,1)
    xx = np.hstack([X[0,:].reshape(-1,1), np.ones(len(X[0,:])).reshape(-1,1)])
    idx = np.random.choice(xx.shape[0], xx.shape[0], replace=True)
    xx = xx[idx,:]
    y = y[idx,:]
    coeff = np.linalg.inv(np.dot(xx.T, xx)).dot(xx.T).dot(y) # beta
    return coeff, xx, y


scores = np.array([1.2,    3.6,  2.7,   4.8,   3.2,   5.9,  6.3,  7.1,  7.5,  9])
N1 = np.array([-0.2, -0.25, -0.5, -0.45, -0.62, -0.83, -0.6, -0.9, -0.7, -1])
X = np.vstack([scores,N1])
plt.scatter(X[0], X[1])
plt.xlabel('Scores')
plt.ylabel('N1, uV')
plt.title('Data from reference ads (N1 vs. scores)')
```

## And the measured N1 response for the target data happens to be -0.65 $\mu V$. We fit a line through the data obtained for reference ads and make a prediction for the target ad based on measured N1:

```python
N1_target = -0.65

plt.scatter(X[0], X[1])
slope, bias = fit(X)
x = np.linspace(min(scores), max(scores), 100)
y = slope*x + bias
plt.plot(x,y)
pred_score = (N1_target-bias)/slope
plt.axhline(y=N1_target, linewidth=1, color='r', alpha = 0.15)
plt.text(1,N1_target+0.01, 'N1_target',rotation=0)
plt.axvline(x=pred_score, linewidth=1, color='r', alpha = 0.15)
plt.text(pred_score-0.3, -0.75, 'Predicted score',rotation=90)
plt.xlabel('Scores')
plt.ylabel('N1, uV')
```

## Then we determine a bootstrapped 95-percent confidence interval:


```python
N1 = -0.65
slope, bias = fit(X)
score = (N1-bias)/slope
print('predicted score:', score[0])

# let's find a 95% bootstrapped confidence interval:
N = 10000
perm_score = np.zeros((N,1)).flatten()
for i in range(N):
    coeff,_,_ = permute(X)
    perm_score[i] = (N1-coeff[1])/coeff[0]
ax = plt.hist(perm_score, bins=np.linspace(0, 10, 50))
q = np.quantile(perm_score, [0.025, 0.975])
plt.axvline(x=q[0], linewidth=1, color='r')
plt.axvline(x=q[1], linewidth=1, color='r')
plt.title('**Bootstrapped 95%-confidence interval:**'+ str(np.round(q,3)))
```

## So, again, the we can say with 95 percent confidence that our prediction for the score of the target ad is:

```python

N1_target = -0.65

ax = plt.scatter(X[0], X[1])
slope, bias = fit(X)
x = np.linspace(min(scores), max(scores), 100)
y = slope*x + bias
plt.plot(x,y)
pred_score = (N1_target-bias)/slope
plt.axhline(y=N1_target, linewidth=1, color='r', alpha = 0.15)
plt.text(1,N1_target+0.01, 'N1_target',rotation=0)
plt.axvline(x=pred_score, linewidth=1, color='r', alpha = 0.15)
plt.text(pred_score-0.3, -0.75, 'Predicted score',rotation=90)
plt.xlabel('Scores')
plt.ylabel('N1, uV')
plt.fill_betweenx(plt.gca().get_ylim(), q[0], q[1], facecolor='green', alpha=0.5)
```

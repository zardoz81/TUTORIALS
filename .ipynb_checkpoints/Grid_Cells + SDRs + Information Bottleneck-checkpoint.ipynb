```python
import math
from scipy.stats import norm
import matplotlib.mlab as mlab

def sdr(n,w):
    X = np.zeros((n,)).astype('int')
    X[np.random.choice(n, w, replace=False)] = 1
    return X

def prob(n,w):
    return math.factorial(n)/(math.factorial(w)*math.factorial(n-w))

def getmax(X_):
    X__ = conv2d(X_, gk.real)
    coord = np.where(X__==X__.max())
    return coord[1], coord[0]

def getprob(n,w,b):
    """_ compute the prbability of a complete overlap due to chance 
    w - SDR population (number of on bits
    n - SDR length number of bits
    b - number of on bits in the overlapping set (or number of on bits in a random noise set)
    _"""
    
    A = 1
    for i in range(b):
        A *= (w-i)/(n-i)
    return A * prob(w,b)

getprob(256,4,4) # get the probability that there will be at least a 3-bit overlap (on-bits) between an SDR 
                # with a population of 4 and a length of 16 and random SDR of the same length and population
```

```python
HH = []
N = 100
n = 16
w0 = 4
w1 = 4
b = 3
for j in range(1000):
    H = []
    for i in range(N):
        if np.logical_and(sdr(n,w0), sdr(n,w1)).sum()>=b:
            H.append(1)
        else:
            H.append(0)
    HH.append(np.sum(H)/N)


plt.axvline(getprob(16,4,b), c='r', label='Theoretical probability: {:.4f}'.format(getprob(16,4,b)))

n, bins, patches = plt.hist(HH, density=True, alpha=0.6, color='g')
mu, std = norm.fit(HH)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2, label='Gaussian fit to the data \n(random trials)')
plt.legend()
```

## Exact probability of FP vs. Numenta's formula
$W$ is the number of on-bits

$M$ is the number of vectors stacked to get the union

$L$ is the length of the vectors (and the union, of course)

$$\sum_{k=0}^W(-1)^k\binom{W}k\left(\frac{\binom{L-k}{W}}{\binom{L}W}\right)^M$$

```python

def FP(L,W,M):
    """ 
    Exact probability of a false positive match 
    https://math.stackexchange.com/questions/3578428/probability-of-a-match-between-a-random-binary-vector-and-a-union-of-random-bina/3578766#3578766
    """
    fp = 0
    for k in range(W+1):
        fp += (-1)**k * prob(W,k) * (prob(L-k,W)/prob(L,W))**M
    return fp

W = 2
M = 2
L = 4
print('Exact p of false positive {}'.format(FP(L,W,M)))

## NUMENTA's formula (not exact, but close)
""" See the details at https://numenta.com/assets/pdf/biological-and-machine-intelligence/BaMI-SDR.pdf """

s = W/L
print((1 - (1-s)**M)**W)
```

$F(x,y)$ - is the revenue that depends on the price of steel and labor

}

## Conditional Entropy
Conditional entropy is the amount of information (i.e. number of bits) needed to describe a codeword $\tilde x \in {\tilde X}$  given that the value of the input variable $x \in X$ is known. The entropy goes to zero if there's a unique mapping between each $x$ and $\tilde x$

## Average Volume

$$\boxed{
2^{H(X|\tilde X)}
}$$

is the number of elements in $X$ that are mapped to the same codeword in $ \tilde X$ 

```python
Px = np.array([0.25,0.25,0.25,0.25])
H = 0
k = 0.999999
pxx = np.eye(4)*np.diag(np.repeat(k, 4)) + np.ones((4,4))*(1-k)/3 - np.diag(np.repeat((1-k)/3, 4)).astype('float32')

plt.figure(figsize=(5,1))
table = plt.table(cellText=np.round(pxx, 3),
          rowLabels=['$x_1$', '$x_2$', '$x_3$', '$x_4$'],
          colLabels=['$ \~{x}_1$', '$\~{x}_2$', '$\~{x}_3$', '$\~{x}_4$'],
          loc='top')
plt.gca().axis('off')
table.scale(1,2)
table.set_fontsize(14)

for x in range(4):
    for x_ in range(4):
        H -= Px[x] * pxx[x_, x] * np.log2(pxx[x_, x])
print('Conditional entropy: {:.5f}'.format(H))
print('Average volume of elements of X mapped to the same element of X_tilde: {:.2f}'.format(2**H))
```

## Now the same thing, but the softest partitioning possible
#### I.e. there is no unambiguous correspondence between $x_i$ and $\tilde{x}_i$, and each $x_i$ corresponds to an average of 4 $\tilde{x}_i$

```python
Px = np.array([0.25,0.25,0.25,0.25])
H = 0
k = 0.25
pxx = np.eye(4)*np.diag(np.repeat(k, 4)) + np.ones((4,4))*(1-k)/3 - np.diag(np.repeat((1-k)/3, 4)).astype('float32')

plt.figure(figsize=(5,1))
table = plt.table(cellText=np.round(pxx, 3),
          rowLabels=['$x_1$', '$x_2$', '$x_3$', '$x_4$'],
          colLabels=['$ \~{x}_1$', '$\~{x}_2$', '$\~{x}_3$', '$\~{x}_4$'],
          loc='top')
plt.gca().axis('off')
table.scale(1,2)
table.set_fontsize(14)

for x in range(4):
    for x_ in range(4):
        H -= Px[x] * pxx[x_, x] * np.log2(pxx[x_, x])
print('Conditional entropy: {:.5f}'.format(H))
print('Average volume of elements of X mapped to the same element of X_tilde: {:.2f}'.format(2**H))
```

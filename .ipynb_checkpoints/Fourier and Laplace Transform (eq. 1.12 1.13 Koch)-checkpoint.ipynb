## Laplace

Suppose we have a differential equation:

$$
\tau \frac{\mathbf{d}v(t)}{\mathbf{d}t} = - v(t) + R \delta(t)
$$
and we want to transform it to Fourier space.

We can use Fourier transform (pain in the ass) or Laplace transform (cool) using the properties (see Wikipedia):

$$
f'(t)) = sF(s)-f(0^{+})
$$
and
$$
af(t) + bg(t) = aF(s) + bG(s)
$$

$$
\mathcal{L} \{ \frac{\mathbf{d}v(t)}{\mathbf{d}t} \}(s) = \tau s V(s) - v(0) = -V(s) + R \cdot 1
$$

$$
\tau s V(s) + V(s) = R
$$

$$
V(s)(\tau s + 1) = R
$$

$$
V(s) = \frac {R}{\tau s + 1}
$$

and since by definition $ s = \sigma + i\omega $ we can write:

$$
\tilde V(f) = \frac {R}{i\omega \tau  + 1}
$$

or 

$$
\boxed{
\tilde V(f) = \frac {R}{1 + i 2 \pi f \tau}
}
$$

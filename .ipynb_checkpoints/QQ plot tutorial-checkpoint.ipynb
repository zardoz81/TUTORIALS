# QQ plot tutorial

If you want to check if the data $X$ fits a particular distribution, **not necessarily a normal distribution**, you can take data $Y$ with some distribution (call it reference distribution), compute quantiles say (10th, 20th, ... 90th) for data X and data $Y$, plot the values of these quantiles and see how good the fit is. Of the fit is almost perfect, you know that data $X$ has almost the same distribution as the reference distribution. ** SEE stat quest on YouTube for more**

```python
import numpy as np
import matplotlib.pyplot as plt

n = 100000

x = np.random.randn(n).flatten()
x = (x-np.mean(x))/np.std(x)
y = np.random.chisquare(5,n).flatten()
y = (y-np.mean(y))/np.std(y)

quantiles = np.arange(0.1,0.9, step=0.05)
x_ = np.quantile(x, quantiles)
y_ = np.quantile(y, quantiles)

fig = plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.hist(x, 20)
plt.title('Normal distribution')

plt.subplot(1,3,2)
plt.hist(y, 20)
plt.title('Chi-squared, df=5')

plt.subplot(1,3,3)
plt.scatter(x_, y_)
plt.plot(np.array([-1,1]), np.array([-1,1]), 'r')
plt.xlabel('Quantiles of data X')
plt.ylabel('Quantiles of data Y')
```

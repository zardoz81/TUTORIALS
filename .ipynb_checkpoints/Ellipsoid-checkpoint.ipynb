```python
import matplotlib.pyplot as plt
import numpy as np

def ellipsoid(theta):
    a = 0.25
    b = 1
    r = a*b/np.sqrt((b*np.cos(theta))**2 + (a*np.sin(theta))**2)
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    return x, y, r
    
theta = np.linspace(0, 2*np.pi, 1000)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(ellipsoid(theta)[0],ellipsoid(theta)[1])
plt.ylabel('$ \sin \theta $')
plt.xlabel('$ \cos \theta $')
plt.xlim(-1,1)
plt.ylim(-1,1)

plt.subplot(1,2,2)
plt.plot(theta, ellipsoid(theta)[1])
plt.title('Ellipsoid', weight='bold')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T, fs, f = 4, 100, 5
t = np.linspace(0,T, fs*T)
s = np.sin(2*np.pi*f*t) # stimulus
r = s + np.random.randn(fs*T,)*0.400 # resonse = stimulus + ε
plt.figure(figsize=(17,5))
plt.plot(t,s, t, r)
```

```python
# OR YOU CAN CREATE A CATEGORICAL VECTOR OF STIMULI, AND A NUMERICAL RESONSE VECTOR (FIRING RATE VALUES):
s = np.random.randint(10, size=(300,)) # stimulus
r = s + np.random.randn(300,)*0.21 # resonse = stimulus + ε
plt.figure(figsize=(17,5))
plt.plot(s)
plt.plot(r)
```

```python
def get_jd(lo, hi, granularity, Q, P):
    xedges = np.linspace(lo,hi,granularity).tolist()
    yedges = np.linspace(lo,hi,granularity).tolist()
    JD,_,_ = np.histogram2d(Q,P, bins=(xedges, yedges), density=False)
    JD = JD/np.sum(JD)
    return JD

def get_entropy(JD, var):
    # entropy of var #
    P = JD.sum(axis=var)
    H = 0
    for i in range(len(P)):
        if P[i]>0:
            H += -P[i]*np.log(P[i])
        else:
            pass
    return H

def get_joint_entropy(JD):
    # joint entropy
    JE = 0
    for i in range(JD.shape[0]):
        for j in range(JD.shape[1]):
            if JD[i,j]>0:
                JE += -JD[i,j]*np.log(JD[i,j])
            else:
                pass
    return JE

def MI(lo, hi, granularity, Q, P):
    JD = get_jd(lo, hi, granularity, Q, P)
    plt.imshow(JD)
    JE = get_joint_entropy(JD)
    H1 = get_entropy(JD,0)
    H2 = get_entropy(JD,1)
    return (H1 + H2 - JE)/np.sqrt(H1+H2)

print(MI(-30,30,100,s,r))    
```

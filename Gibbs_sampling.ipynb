We randomly choose $y$ (in the code we start by x, but it doens't change anything). Having chosen (or fixed) y at some value $y^{(j-1)}$, we select only the values of $x$, whose corresponding y's are equal to $y^{(j-1)}$. Thus, we get the conditional distribution $P \big(x|y^{(j-1)}\big)$. Then, we sample some $x$ from this conditional distribution:

$$
x^{(0)} \sim P \big(x|y^{(j-1)}\big)
$$
Then we create a condional distribution of $y$ given $x=x^{(0)}$
$$
y^{(1)} \sim P \big(y|x^{(0)}\big)
$$

and continue the same procedure for several steps. This allows us to sample a from very high-dimensional distributions efficiently.

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Gibbs sampling
"""

import matplotlib.pyplot as plt
import numpy as np


cov = np.array([[1.0 ,0.5], [0.5, 1.0]])
mean = np.array([0, 0])
x, y = np.random.multivariate_normal(mean, cov, 7000).T

x = np.round(x, 2)
y = np.round(y, 2)
plt.plot(x, y, 'x', label='Original multivariate distribution')
plt.xlim(-4, 4)
plt.ylim(-4, 4)



x_s = np.random.choice(x, size=1, replace=True)
X_, Y_ = [], []

for n in range(6000):
    # record the randomly sampled x_s
    X_.append(x_s)
    # get p(y|x)
    dist_y_x = y[np.nonzero([x==x_s])[1]]
    # sample y from this conditional distribution
    y_s = np.random.choice(dist_y_x, size=1, replace=True)
    # record the randomly sampled y_s
    Y_.append(y_s)
    
    # get p(x|y)
    dist_x_y = x[np.nonzero([y==y_s])[1]]
    # sample x fomr this conditional distribution
    x_s = np.random.choice(dist_x_y, size=1, replace=True)
    
plt.plot(X_, Y_, 'x', c='r', label='Gibbs-sampled distribution')
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.legend()

```

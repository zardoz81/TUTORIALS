```python
N = 1000
X = torch.randn(N, dtype=torch.double)
Y = torch.randn(N, dtype=torch.double)
```

```python
plt.figure(figsize=(15,4))
plt.subplot(1,2,1)
plt.hist(X)
plt.title('X')
plt.subplot(1,2,2)
plt.hist(Y)
plt.title('Y')
```

```python
torch.distributions.kl.kl_divergence(X, Y)
```

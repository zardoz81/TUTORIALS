## Here we will create a variational autoencoder that does not combine cross-entropy and KL-divergence losses into one, but passes them separately into the model compiler. For more information and KL-divergence derivation, see a separate Tutorial on Variational Autoencoder

```python
# make sure you don't hog all the video memory
import tensorflow as tf
config = tf.ConfigProto()
config.gpu_options.allow_growth = True
sess = tf.Session(config=config)
from keras import backend as K
K.set_session(sess)
###################################

# imports:
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from keras import metrics, objectives, losses
from keras.layers import Input, Dense, Lambda
from keras.models import Model
from keras import backend as K
from keras import metrics
from keras.datasets import mnist
```

```python
original_dim = 784
latent_dim = 2
intermediate_dim = 256
batch_size = 100
epsilon_std = 1.0
```

```python
# sampling, sampling2 and sampling3 are three different impolementations that do the same thing:

def sampling(args):
    """
    This function samples 2-D points from a binomial normal distribution with given mean (2d) and std (2d)
    """
    z_mean, z_log_sigma = args

    epsilon = K.random_normal(shape=(latent_dim,), mean=0., stddev=epsilon_std)
    return z_mean + K.exp(z_log_sigma) * epsilon #!!!! WE NEED TO GET 2-VECTORS FROM A WHITE 2D NORMAL DISTRIBUTION
    # THE 2D NORMAL DISTRIBUTION IS PARAMETERIZED A 2-MEAN AND AND 2-LOG-SIGMA. So here we just take a random 
    # 2-vector (with mean 0 and std 1), then we offset these points from (0,0) by 
    # some std K.exp(z_log_sigma) * epsilon

def sampling2(args):
    """
    For each batch the encoder has two outputs:
    z_mean of size (batch_size, 2)
    z_log_sigma of size (batch_size, 2)
    These are parameter pairs of bivariate normal distributions, from which the sampling2 function 
    will draw 2D-points (one 2D point per each row of the batch). These 2-vectors are fed into the DECODER.
    In other words, we draw these 2D points from bivariate normal distributions P(z|x),
    each of which is conditioned on its respecitve input image x.
    """
    sample = []
    z_mean, z_log_sigma = args
#     print('z_mean.shape: ', z_mean.shape)
#     print('z_log_sigma.shape: ', z_log_sigma.shape)
    for i in range(batch_size):
        line = []
        dim1 = K.random_normal(shape=(1,), mean=z_mean[i,0], stddev=K.exp(z_log_sigma[i,0]))
        dim2 = K.random_normal(shape=(1,), mean=z_mean[i,1], stddev=K.exp(z_log_sigma[i,1]))
        line.append(dim1)
        line.append(dim2)
        line = K.stack(line, axis=1)
#         print(line.shape)
        sample.append(line)
    return K.reshape(K.stack(sample, axis=0), shape=(batch_size, 2))

def sampling3(args):
    """
    PERFORMANCE-OPTIMIZED VERSION OF sampling2 (see above, also a demo above)
    For each batch the encoder has two outputs:
    z_mean of size (batch_size, 2)
    z_log_sigma of size (batch_size, 2)
    These are parameter pairs of bivariate normal distributions, from which the sampling2 function 
    will draw 2D-points (one 2D point per each row of the batch). These 2-vectors are fed into the DECODER.
    In other words, we draw these 2D points from bivariate normal distributions P(z|x),
    each of which is conditioned on its respecitve input image x.
    """
    z_mean, z_log_sigma = args
    return K.random_normal(shape=(1,latent_dim), mean=z_mean, stddev=K.exp(z_log_sigma))


def combine(args):
    """
    We have to use this function as a hack for the following reason. We need to create a custom loss (to compute KL divergence between
    Q(z|x) and P ~ N(1, I) ). Because to compute KL divergence we need two tensors output by the two layers `z_mean` and `z_log_mean`, 
    we would have fed them into the loss function as a list of two tensors. However, in Keras, you can only feed the ouput of one layer into
    a loss function. As a workaround, we create a custom layer (Lambda layer) that takes a list of two tensors (one from z_mean, one from z_log_mean),
    and stack that list into a tensor of shape (?, 4). The output of this Lambda layer is a single tensor and is the input to the kl_loss
    loss function.
    """
    out = K.stack(args, axis=1)
    out = K.reshape(out, shape=(-1,4))
#     print('out.shape:',out.shape)
    return out
    
def xent_loss(targets, predictions):
    """
    Note that the compiler outputs two tensors of which targets is the first, and predictions is the sencond!!! These tensors are fed into
    the loss functions listed in the compile function.
    """
    return original_dim * objectives.binary_crossentropy(targets, predictions)

def kl_loss(dummy_targets, combined_z_std):
    print(combined_z_std.shape)                 # you can review the shapes of tensors during the compilation of the model
    z_mean = combined_z_std[:,0:2]
    z_log_sigma = combined_z_std[:,2:4]         # unpack the list (consisting of layer 1 (z_mean) and layer 2 (z_log_sigam) outputs)
    return -0.5 * K.sum(1 + z_log_sigma - K.square(z_mean) - K.exp(z_log_sigma), axis=-1) # for derivation, see the separate tutorial

# x = Input(shape=(batch_size, original_dim))
x = Input(shape=(original_dim,))
h = Dense(intermediate_dim, activation='relu')(x)
z_mean = Dense(latent_dim, name='z_mean')(h)
z_log_sigma = Dense(latent_dim, name='z_log_sigma')(h)

z = Lambda(sampling, output_shape=(latent_dim,), name='sampling')([z_mean, z_log_sigma]) # use sampling, sampling2 or sampling3
combined = Lambda(combine, output_shape=(4,), name='combined')([z_mean, z_log_sigma])


decoder_h = Dense(intermediate_dim, activation='relu')
decoder_mean = Dense(original_dim, activation='sigmoid')
h_decoded = decoder_h(z)
x_decoded_mean = decoder_mean(h_decoded)
```

```python
# end-to-end autoencoder
vae = Model(x, [x_decoded_mean, combined])

# encoder, from inputs to latent space
encoder = Model(x, z_mean)

# generator, from latent space to reconstructed inputs
decoder_input = Input(shape=(latent_dim,))
_h_decoded = decoder_h(decoder_input)
_x_decoded_mean = decoder_mean(_h_decoded)
generator = Model(decoder_input, _x_decoded_mean)
vae.summary()
vae.compile(optimizer='rmsprop', loss=[xent_loss, kl_loss])
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.
x_train = x_train.reshape((len(x_train), np.prod(x_train.shape[1:])))
x_test = x_test.reshape((len(x_test), np.prod(x_test.shape[1:])))
```

```python
"""
Take a look at what the autoencoder gives us BEFORE TRAINING:
"""
# x_test_encoded = encoder.predict(x_test, batch_size=batch_size)
plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
x_test_encoded = encoder.predict(x_test)
plt.scatter(x_test_encoded[:, 0], x_test_encoded[:, 1], c=y_test)
plt.colorbar()
# plt.show()


plt.subplot(1,2,2)
# display a 2D manifold of the digits
n = 25  # figure with 15x15 digits
digit_size = 28
figure = np.zeros((digit_size * n, digit_size * n))
# we will sample n points within [-15, 15] standard deviations
grid_x = np.linspace(-2, 2, n)
grid_y = np.linspace(-2, 2, n)

for i, yi in enumerate(grid_x):
    for j, xi in enumerate(grid_y):
        z_sample = np.array([[xi, yi]]) * epsilon_std
        x_decoded = generator.predict(z_sample)
        digit = x_decoded[0].reshape(digit_size, digit_size)
        figure[i * digit_size: (i + 1) * digit_size,
               j * digit_size: (j + 1) * digit_size] = digit
plt.imshow(figure)
```

# Be careful: in Keras, the number of loss functions should match the number of outputs from the model and length of the list of targets. If you actually need external targets to compute loss, just create dummy tensors (for training and validation) of size equal to the relevant output. 

## -----------------------
## Another important thing. In Keras, if you assemble a model from initialized layers (like in this tutorial) the weights in these layers are shared. I.e. if you train "vae" in this example, the the weight of the "encoder" are the same (in corresponding layers). We'll verify it below.

```python
dummy2_train = np.zeros([60000,4], dtype='float32')
dummy2_test = np.zeros([10000,4], dtype='float32')

h = vae.fit(x_train, [x_train, dummy2_train],
        shuffle=True,
        epochs=20,
        verbose = 1,
        batch_size=batch_size,
        validation_data=(x_test, [x_test, dummy2_test]))
```

```python
KL_div = h.history['combined_loss']
KL_div = KL_div/KL_div[0]
x_entropy = h.history['dense_3_loss'] # CHANGE THIS TO MATCH THE LAYER NAME (SEE TRAINING PROGRESS ABOVE)
x_entropy = x_entropy/x_entropy[0]
plt.plot(x_entropy)
plt.plot(KL_div)
plt.title('Normalized losses')
ax = plt.legend(['Cross-entropy','KL-divergence'])
```

## Now we can see that the "z_mean" layer of the "vae" model and the final layer of the "encoder" model produce the same values if you forward the training tensor through them:

```python
inp = vae.input                                           # input placeholder
output = vae.layers[2].output # get the output tensor from layer #2
print(vae.layers[2].name)     # get the layer's name
functor = K.function([inp, K.learning_phase()], [output]) # define teh function that will get the activations of our layer

# forward x_test through the vae model and get the activations at the set layer:
layer_outs = functor([x_test, 1.])

plt.scatter(layer_outs[0][:, 0], layer_outs[0][:, 1], c=y_test)
plt.colorbar()
```

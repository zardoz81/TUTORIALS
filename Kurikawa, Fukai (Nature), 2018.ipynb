```python
import numpy as np
import matplotlib.pyplot as plt
```

$$
r_i = \frac{R_{max}}{L}e^{\frac{\big(-\frac{i}{N_{in}}-0.1(k+1)\big)^2}{2(\sigma L)^2}}
$$

```python
R_max = 120
N_in = 200
sigma = 0.05
L = 0.4
i = np.arange(1,201).astype('float').reshape(-1,1) # neuron index
k = np.arange(1,8).astype('float').reshape(1,-1) # stimulus sound frequency
```

```python
r_i = (R_max/L)*np.exp(-((i/N_in)-0.1*(k+1.0))**2/(2*(sigma/L)**2))
```

```python
plt.plot(i, r_i)
plt.legend({'13', '12.5', '12', '11.5', '11', '10.5', '10'}, title='Stimulus \nfrequency')
plt.ylabel('Spiking rate of ith neuron, Hz')
plt.xlabel('Neuron\'s index, i')
plt.title('Tuning curves of 200 neurons')
```

## deep dream tutorial

```python
model = 1
add_softmax = False
cuda = 1
DO = 0.2

from keras.utils import to_categorical
from keras import optimizers

import scipy
import numpy as np

import matplotlib.pyplot as plt

import torch
print(torch.__version__)
from torch.autograd import Variable
import torch.optim as optim
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision

# import the model and switch off dropout and fix the parameters:
if cuda==1:
    torch.cuda.set_device(1)
    torch.set_default_tensor_type('torch.cuda.FloatTensor')
if cuda==0:
    torch.set_default_tensor_type('torch.FloatTensor')

if model==1:
    vgg16 = torchvision.models.vgg16(pretrained=True)
    list(vgg16.children())[1][2].p = DO
    list(vgg16.children())[1][5].p = DO
    from keras.applications.vgg16 import decode_predictions
    for i in vgg16.parameters():
        i.requires_grad = False
    if add_softmax==True:
        vgg16.classifier.add_module('7', nn.Softmax(dim=1))
        print('Softmax added')
    else:
        print('No Softmax added')
    print('Model: ', model)

if model==2:
    iv3 = torchvision.models.inception_v3(pretrained=True)
    from keras.applications.inception_v3 import decode_predictions
    for i in iv3.parameters():
        i.requires_grad = False
    print('Model: ', model)


def scaler(vector, ceiling):
    vector = vector.float()
    min_v = torch.min(vector).float()
    range_v = torch.max(vector) - min_v
    if range_v > 0:
        normalised = (vector - min_v) / range_v * ceiling
    else:
        normalised = torch.zeros(vector.size())
    return normalised

data_dir = '/home/amplifier/home/NEW_DL/img/'

data_transforms = torchvision.transforms.Compose([
        # Data augmentation is a good practice for the train set
        # Here, we randomly crop the image to 224x224 and
        # randomly flip it horizontally. 
        torchvision.transforms.Resize((224, 224)),
        torchvision.transforms.ToTensor(),
    ])

train_dataset = torchvision.datasets.ImageFolder(
    root=data_dir,
    transform=data_transforms
)
train_loader = torch.utils.data.DataLoader(
    train_dataset,
    batch_size=3,
    num_workers=0,
    shuffle=True
)
print('Dropout:', DO)
```

```python
torch.cuda.empty_cache()
# get one image:
bat = list(train_loader)[0]
if model==1:
    dat = bat[0][0,:,:,:].view(1,3,224,224)
#     dat = torch.randn(1,3,224,224) # if you want to use random noise as input to the model 
#     target = torch.tensor([953], dtype=torch.long) # gradient descend to pineapple
    target = torch.tensor([torch.argmax(vgg16(dat.cuda()))]) # target: what the network 'sees' in the dat image (to gradient ascend from)

if model==2:
    dat = bat[0][0:2,:,:,:].view(2,3,224,224)
    target = torch.empty(2, dtype=torch.long).random_(1000)
if cuda==1:
    dat = dat.cuda()
if cuda==0:
    dat = dat.cpu()

plt.imshow(dat[0,:,:,:].cpu().detach().numpy().transpose(1,2,0))
plt.title(decode_predictions(to_categorical(target.cpu().detach().numpy(), 1000))[0][0][1])
dat1 = dat.cpu().detach().numpy()
```

```python
criterion = nn.CrossEntropyLoss()

if cuda==1:
    x = Variable(dat, requires_grad=True).cuda()
    target = target.cuda()
else:
    x = Variable(dat, requires_grad=True).cpu()
    target = target.cpu()

if model==1:
    output = vgg16(x)
    loss = criterion(output, target)
    
if model==2:
    output = iv3(x)
    loss = criterion(output[0], target)

h = []
l1_loss = []
optimizer = optim.Adam([x])
output_init = vgg16(x).cpu().detach().numpy().flatten()
```

```python
for j in range(100):
    target = torch.tensor([torch.argmax(vgg16(dat.cuda()))]) # tell the model to enhance what it sees
    print('I see:', decode_predictions(to_categorical(target.cpu().detach().numpy(), 1000))[0][0][1])
    for i in range(3):
        output = vgg16(x)
        l1 = torch.norm(output,1)
        loss = criterion(output, target) + 0.1 * l1
        l1_loss.append(l1.item())
        h.append(loss.item())
        loss.backward(retain_graph=True)
        # either:
        x.grad.data = -x.grad.data  # for gradient ascent (remove the - for descent)
        optimizer.step()
        optimizer.zero_grad()
        # or:
        # x.data.sub_(-x.grad.data*0.01) # for gradient ascent (remove the - for descent)
        # x.grad.data.zero_()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(dat1[0,:,:,:].transpose(1,2,0))
plt.title(decode_predictions(output_init.reshape(1,1000))[0][0][1])

plt.subplot(1,3,2)
plt.imshow(scaler(x[0,:,:,:], 1).cpu().detach().numpy().transpose(1,2,0))
plt.title(decode_predictions(vgg16(x).detach().cpu().numpy())[0][0][1])

plt.subplot(1,3,3)
plt.plot(h)
plt.title('Target: {},\n Model says: {}'.format(decode_predictions(to_categorical(target.cpu().detach().numpy(), 1000))[0][0][1],
                                          decode_predictions(output.cpu().detach().numpy())[0][0][1]))
plt.xlabel('switch off dropout, you get smooth learning curve')
```

## you can do gradient ascent not from the final layer, but from the last layer of the feature extractor. To enhance either what the feature extractor "sees" in the picture, or in pure noise.

```python
def get_reference_im():
    ds = torchvision.datasets.ImageFolder(
    root='/home/amplifier/home/NEW_DL/Ref_im/',
    transform=data_transforms)
    return list(torch.utils.data.DataLoader(ds))[0][0]
reference = get_reference_im().cuda()
dat = bat[0][0,:,:,:].view(1,3,224,224).cuda()

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(reference[0,:,:,:].detach().cpu().numpy().transpose(1,2,0))
plt.subplot(1,3,2)
plt.imshow(dat[0,:,:,:].detach().cpu().numpy().transpose(1,2,0))
```

```python
criterion = nn.MSELoss()

# initialize the image to deep dream on:
# x = Variable(torch.randn(1,3,224,224), requires_grad=True).cuda() # either random noise
x = Variable(dat, requires_grad=True).cuda() # or real image

optimizer = optim.Adam([x])

# make a network (here we can just simply take the layers we want from a bigger model):
small = nn.Sequential(*list(vgg16.features.children())[0:22])

h = []
l1_loss = []
```

```python
for j in range(1):
    # we ask the network what it 'sees' in the image:
#     target = small(dat.cuda())
    plt.imshow(dat.detach().cpu().numpy().squeeze().transpose(1,2,0))
    
    # target = torch.tensor([torch.argmax(vgg16(dat.cuda()))]) # tell the model to enhance what it sees
    # target = vgg16.features(dat.cuda())
    # we can even decode what the network sees if we use the last layer to do gradient ascent from:
    # print('I see:', decode_predictions(to_categorical(target.cpu().detach().numpy(), 1000))[0][0][1])
    
    # or we can give the network a reference image to gradient descend on:
    target = small(reference)
    
    for i in range(300):
#         output = vgg16.features(x)
        output = small(x)
        l1 = torch.norm(output,1)
        loss = criterion(output, target) + 0.01 * l1
        l1_loss.append(l1.item())
        h.append(loss.item())
        loss.backward(retain_graph=True)
        # either:
        x.grad.data = -x.grad.data  # for gradient ascent (remove the - for descent)
        optimizer.step()
        optimizer.zero_grad()
        # or:
        # x.data.sub_(-x.grad.data*0.01) # for gradient ascent (remove the - for descent)
        # x.grad.data.zero_()
```

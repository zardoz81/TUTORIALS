```python
import numpy as np
import matplotlib.pyplot as plt

N = 13
Dxhat = np.linspace(-2,2,N)
Dx = np.linspace(-2,2,N)

L = np.zeros((N,N))
for i in range(N):
    for j in range(N):
        L[i,j] = np.max([0, 1+Dxhat[i]]) + np.max([0, 1-Dx[j]])
plt.figure(figsize=(7,7))
plt.imshow(L)
plt.yticks(np.arange(N), np.round(Dxhat,1))
plt.xticks(np.arange(N), np.round(Dx,1))
plt.ylabel('$D_\hat{x}$', fontsize=18)
plt.xlabel('$D_{x}$', fontsize=18)
```

As you can see from the heatmap above, the $L_{DSC}$ loss becomes smaller as $D_{\hat x}$ becomes smaller and as $D_{x}$ becomes larger. In other word, as realism scores for fakes get smaller while getting larger for reals.

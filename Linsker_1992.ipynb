```python
import numpy as np
import matplotlib.pyplot as plt

s = np.random.randn(10,10000)
n = np.random.randn(10,10000)

print(np.mean(s))
print(np.mean(n))
```

```python
a = (s + n).dot((s + n).T)/10000
b = s.dot(s.T)/10000 + n.dot(n.T)/10000
c = n.dot(s.T)/10000
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(a)
plt.title('Sumsq: {:.3f}'.format(np.sum(a**2)))
plt.colorbar()
plt.subplot(1,3,2)
plt.imshow(b)
plt.title('Sumsq: {:.3f}'.format(np.sum(b**2)))
plt.colorbar()
plt.subplot(1,3,3)
plt.imshow(c)
plt.title('Sumsq: {:.3f}'.format(np.sum(c**2)))
plt.colorbar()
```

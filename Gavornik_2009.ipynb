```python
x = np.random.choice(2, 10)
y = np.logical_not(x) * 1

plt.figure()
plt.plot(x)
plt.plot(y)

I1 = x/np.linalg.norm(x)
I2 = y/np.linalg.norm(y)
l1 = 2
l2 = 3

XX = l1* np.outer(I1,I1) + l2* np.outer(I2,I2)
u, s, vh = np.linalg.svd(XX) # the columns of u are the eigenvectors of A A^H
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.imshow(XX)
plt.subplot(2,2,2)
plt.plot(s, 'd')
plt.subplot(2,2,3)
plt.plot(u[:,0])
plt.plot(u[:,1])
plt.subplot(2,2,4)
plt.imshow(u)
```

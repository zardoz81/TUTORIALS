```python
import numpy as np
import matplotlib.pyplot as plt


Cp = 0.00525
Cd = 0.0105
td = 20
tp = 40

d = np.linspace(-200,200, 400)

def G(d):
    return Cp * np.exp(-np.abs(d)/tp) - Cd * np.exp(-np.abs(d)/td)

plt.plot(d, G(d))
plt.grid()

plt.plot(d, Cp * np.exp(-np.abs(d)/tp))
plt.plot(d, Cd * np.exp(-np.abs(d)/td))
```
